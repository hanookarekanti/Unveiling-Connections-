```python
# Step 1: Import libraries
import pandas as pd
import os
import re
import unicodedata
import random
import spacy
from collections import defaultdict
import numpy as np
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network
from google.colab import files

# Step 2: Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load spaCy model
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

# Step 3: Create data folder in Colab
if not os.path.exists("data"):
    os.makedirs("data")

# Step 4: Upload the Bhagwad_Gita.csv file to Colab
if not os.path.exists("Bhagwad_Gita.csv"):
    print("Please upload Bhagwad_Gita.csv:")
    uploaded = files.upload()

# Step 5: Define entity mappings, person names, locations, etc.
entity_map = {
    'Arjun': 'Arjuna', 'Krishn': 'Krishna', 'Yudhishtira': 'Yudhishthira',
    'Bheema': 'Bhima', 'Nakul': 'Nakula', 'Sahadev': 'Sahadeva',
    'Draupady': 'Draupadi', 'Kuntee': 'Kunti', 'Madhusudana': 'Krishna',
    'Hrishikesha': 'Krishna', 'Janardana': 'Krishna', 'Vaasudeva': 'Krishna',
    'Vasudeva': 'Krishna', 'Govinda': 'Krishna', 'Hari': 'Krishna', 'Keshava': 'Krishna',
    'Parthasarathi': 'Krishna', 'Yogeshwara': 'Krishna', 'Achyuta': 'Krishna',
    'Narayana': 'Krishna', 'Vishnu': 'Krishna', 'Partha': 'Arjuna',
    'Kaunteya': 'Arjuna', 'Dhananjaya': 'Arjuna', 'Gudakesha': 'Arjuna',
    'Bharatarshabha': 'Arjuna', 'Kiriti': 'Arjuna', 'Savyasachi': 'Arjuna',
    'Vijaya': 'Arjuna', 'Pritha': 'Kunti', 'Devavrata': 'Bhishma',
    'Gangaputra': 'Bhishma', 'Pitamaha': 'Bhishma', 'Acharya Drona': 'Drona',
    'Kripacharya': 'Kripa', 'Veda Vyasa': 'Vyasa', 'Krishna Dvaipayana': 'Vyasa',
    'Radheya': 'Karna', 'Vaikartana': 'Karna', 'Suryaputra': 'Karna',
    'Angaraja': 'Karna', 'Ashwathama': 'Asvatthama', 'Ashvatama': 'Asvatthama',
    'Shikhandini': 'Sikhandi', 'Alambusha': 'Alambusha', 'Bahlika': 'Bahlika',
    'Shrutayudha': 'Shrutayudha', 'Sudakshina': 'Sudakshina', 'Sarvaga': 'Sarvaga',
    'Pradyumna': 'Pradyumna', 'Samba': 'Samba', 'Satyavati': 'Satyavati',
    'Amba': 'Amba', 'Ambika': 'Ambika', 'Ambalika': 'Ambalika',
    'Phalguna': 'Arjuna', 'Jishnu': 'Arjuna', 'Kiritin': 'Arjuna', 'Pandava': 'Arjuna',
    'Madhava': 'Krishna', 'Damodara': 'Krishna', 'Yadava': 'Krishna',
    'Dharmaraja': 'Yudhishthira', 'Ajatasatru': 'Yudhishthira',
    'Panchali': 'Draupadi', 'Krishnaa': 'Draupadi', 'Yajnaseni': 'Draupadi',
    'Dhrishtadyumna': 'Dhrishtadyumna', 'Duryodhan': 'Duryodhana', 'Suyodhana': 'Duryodhana',
    'Kururaja': 'Duryodhana', 'Gandhari-putra': 'Duryodhana', 'Dhritarashtraputra': 'Duryodhana',
    'Bheeshma': 'Bhishma', 'Shantanava': 'Bhishma', 'Gandiva-dhari': 'Arjuna',
    'Vrikodara': 'Bhima', 'Bhimsen': 'Bhima', 'Subhadraa': 'Subhadra', 'Yadavi': 'Subhadra',
    'Balaram': 'Balarama', 'Baladeva': 'Balarama', 'Sankarshana': 'Balarama',
    'Gandhariputra': 'Dushasana', 'Dussasana': 'Dushasana', 'Madriputra': 'Nakula',
    'Madreyi-suta': 'Sahadeva', 'Kaurava': 'Duryodhana', 'Shakuni': 'Shakuni',
    'Gandhara-raja': 'Shakuni', 'Abhimanyu': 'Abhimanyu', 'Saubhadra': 'Abhimanyu',
    'Parikshit': 'Parikshit', 'Janamejaya': 'Janamejaya', 'Ghatotkacha': 'Ghatotkacha',
    'Hidimba-suta': 'Ghatotkacha', 'Babhruvahana': 'Babhruvahana', 'Iravan': 'Iravan',
    'Uloopi-suta': 'Iravan', 'Chitrangada-suta': 'Babhruvahana', 'Vrishasena': 'Vrishasena',
    'Sushena': 'Sushena', 'Uttaraa': 'Uttara', 'Virataputri': 'Uttara',
    'Bhanumati': 'Bhanumati', 'Vrushali': 'Vrushali', 'Hidimbi': 'Hidimbi',
    'Uloopi': 'Ulupi', 'Chitrangada': 'Chitrangada', 'Devika': 'Devika',
    'Karenumati': 'Karenumati', 'Vijayaa': 'Vijaya', 'Ganga': 'Ganga',
    'Shantanu': 'Shantanu', 'Dhritarastra': 'Dhritarashtra', 'Pandoo': 'Pandu',
    'Madri': 'Madri', 'Gandhari': 'Gandhari', 'Devaki': 'Devaki',
    'Vasudev': 'Vasudeva', 'Dushala': 'Dushala', 'Jayadratha': 'Jayadratha',
    'Satyaki': 'Satyaki', 'Yuyutsu': 'Yuyutsu', 'Shalya': 'Shalya',
    'Dhrishtadyumna': 'Dhrishtadyumna', 'Draupada': 'Drupada', 'Virata': 'Virata',
    'Somadatta': 'Somadatta', 'Bhurisrava': 'Bhurisrava', 'Laxman': 'Laxman',
    'Chitrasena': 'Chitrasena', 'Chitravarma': 'Chitravarma', 'Yaudheya': 'Yaudheya',
    'Niramitra': 'Niramitra', 'Suhotra': 'Suhotra', 'Shrutakarma': 'Shrutakarma',
    'Sutasoma': 'Sutasoma', 'Shrutasena': 'Shrutasena', 'Shrutakirti': 'Shrutakirti',
    'Prativindhya': 'Prativindhya', 'Vikarna': 'Vikarna', 'Vidura': 'Vidura'
}


person_names = [
    'Dhritarashtra', 'Sanjaya', 'Duryodhana', 'Drona', 'Pandu', 'Arjuna', 'Bhima',
    'Yudhishthira', 'Nakula', 'Sahadeva', 'Krishna', 'Bhishma', 'Karna', 'Kripa',
    'Asvatthama', 'Vikarna', 'Bhurisrava', 'Drupada', 'Virata', 'Satyaki', 'Dhrishtaketu',
    'Chekitana', 'Purujit', 'Kuntibhoja', 'Saibya', 'Yodhamanyu', 'Uttamaujas', 'Abhimanyu',
    'Subhadra', 'Draupadi', 'Kunti', 'Pritha', 'Sikhandi', 'Dhrishtadyumna', 'Jayadratha',
    'Vyasa', 'Brahma', 'Manu', 'Ikshvaku', 'Vivasvan', 'Narada', 'Kapila', 'Bhrigu', 'Usanas',
    'Vaasudeva', 'Vasudeva', 'Madhava', 'Govinda', 'Hari', 'Keshava', 'Janardana', 'Parthasarathi',
    'Yogeshwara', 'Achyuta', 'Narayana', 'Hrishikesha', 'Madhusudana', 'Vishnu', 'Partha', 'Kaunteya',
    'Dhananjaya', 'Gudakesha', 'Bharatarshabha', 'Kiriti', 'Savyasachi', 'Vijaya', 'Dharma Raj',
    'Ajatashatru', 'Vrikodara', 'Suyodhana', 'Dushasana', 'Bhagadatta', 'Devavrata', 'Gangaputra',
    'Pitamaha', 'Acharya Drona', 'Bharadvaja-putra', 'Kripacharya', 'Veda Vyasa', 'Krishna Dvaipayana',
    'Radheya', 'Vaikartana', 'Suryaputra', 'Angaraja', 'Shalya', 'Ghatotkacha', 'Uttara', 'Kritavarma',
    'Markandeya', 'Angira', 'Vashishtha', 'Gautama', 'Kashyapa', 'Chyavana', 'Vishvamitra', 'Shiva',
    'Mahadeva', 'Rudra', 'Indra', 'Yama', 'Mrityu', 'Agni', 'Varuna', 'Vayu', 'Aditya', 'Surya', 'Chandra',
    'Kubera', 'Hanuman', 'Maruti', 'Anjaneya', 'Garuda', 'Vasuki', 'Shesha', 'Devaki', 'Somadatta',
    'Alambusha', 'Bahlika', 'Shrutayudha', 'Sudakshina', 'Sarvaga', 'Pradyumna', 'Samba', 'Satyavati',
    'Amba', 'Ambika', 'Ambalika','Gandhari', 'Madri', 'Shakuni', 'Balarama', 'Parikshit', 'Janamejaya',
    'Hidimbi','Ulupi', 'Chitrangada', 'Devika', 'Karenumati', 'Vijaya', 'Bhanumati', 'Vrushali',
    'Ganga', 'Shantanu', 'Vidura', 'Dushala', 'Yuyutsu', 'Laxman', 'Chitrasena',
    'Chitravarma', 'Yaudheya', 'Niramitra', 'Suhotra', 'Iravan', 'Babhruvahana',
    'Vrishasena', 'Sushena', 'Shrutakarma', 'Sutasoma', 'Shrutasena', 'Shrutakirti',
    'Prativindhya', 'Kshatradeva', 'Kshatradharman', 'Satyajit', 'Kekaya', 'Brihadkshatra',
    'Jarasandha', 'Rukmini', 'Satyabhama', 'Jambavati', 'Rohini', 'Aniruddha',
    'Charudeshna', 'Saurin', 'Sini', 'Vinda', 'Anuvinda', 'Shishupala', 'Rukmi',
    'Kamsa', 'Ekalavya', 'Srinjaya', 'Dhruva', 'Ugrasena', 'Vasudeva (father)', 'Rohini (mother)'
]


locations = [
    'Kurukshetra', 'Dharmakshetra', 'Hastinapura', 'Dwarka', 'Mathura', 'Kasi', 'Kuru',
    'Panchala', 'Matsya', 'Magadha', 'Bharata', 'Varshneya', 'Ganga', 'Yamuna', 'Indraloka',
    'Swarga', 'Brahmaloka', 'Vaikuntha', 'Dharmakshetra Kurukshetra', 'Kaurava', 'Pandava',
    'Bharatavarsha', 'Ayodhya', 'Gandhara', 'Anga', 'Videha', 'Kuru Jangala', 'Indraprastha',
    'Kamyaka', 'Dvaitavana', 'Khandavaprastha', 'Khandava', 'Virata', 'Kaurava camp', 'Pandava camp',
    'Kurus', 'Panchal', 'Bharat', 'Kosal', 'Madra',
    'Chedi', 'Avanti', 'Sindhu', 'Kekaya', 'Trigarta', 'Pragjyotisha', 'Manipura',
    'Vanga', 'Kalinga', 'Kosala', 'Mithila', 'Surasena', 'Bhoja', 'Nishada',
    'Shalva', 'Kamboja', 'Yadava', 'Saraswati', 'Godavari', 'Narmada', 'Sindhu (river)',
    'Kubera’s palace', 'Yamaloka', 'Varunaloka', 'Vayuloka', 'Agniloka', 'Gokula',
    'Vrindavana', 'Barsana', 'Nandagaon', 'Mount Meru', 'Himalaya', 'Mandara',
    'Gandhamadana', 'Badrikashrama', 'Haridwar', 'Prayaga', 'Vishnupada', 'Champa',
    'Pundravardhana', 'Tamralipta', 'Ujjayini', 'Taxila', 'Pushkalavati', 'Rajagriha',
    'Girivraja', 'Pataliputra', 'Kundina', 'Vidarbha', 'Dwaravati', 'Anarta',
    'Saurashtra', 'Satyapura', 'Kanyakubja', 'Ahichhatra', 'Kampilya', 'Ekacakra',
    'Varnavata', 'Upaplavya', 'Matsya kingdom', 'Saindhava', 'Drupada’s palace',
    'Hidimba’s forest', 'Manasa', 'Alaka', 'Amaravati', 'Nagapura', 'Kailasa',
    'Patala', 'Rakshasa loka', 'Yaksha loka', 'Gandharva loka'
]


relation_keywords = [
    'son', 'sons', 'daughter', 'daughters', 'father', 'fathers', 'mother', 'brother',
    'brothers', 'sister', 'sisters', 'cousin', 'cousins', 'grandfather', 'grandfathers',
    'grandsire', 'grandson', 'grandsons', 'maternal uncle', 'maternal uncles',
    'father-in-law', 'brother-in-law', 'husband', 'wife', 'child', 'children',
    'parent', 'parents', 'offspring', 'sibling', 'siblings', 'brethren', 'son of',
    'daughter of', 'born to', 'born of', 'children of', 'child of', 'kinsman',
    'heir', 'descendant', 'descendants', 'progeny', 'ancestors', 'lineage', 'born',
    'parentage', 'forefather', 'progenitor', 'scion', 'issue', 'begotten', 'sire',
    'ancestry', 'heritage', 'clan', 'tribe', 'household', 'line', 'race', 'blood',
    'grandmother', 'grandmothers', 'granddaughter', 'granddaughters', 'uncle', 'uncles',
    'aunt', 'aunts', 'nephew', 'nephews', 'niece', 'nieces', 'mother-in-law',
    'sister-in-law', 'son-in-law', 'daughter-in-law', 'stepbrother', 'stepbrothers',
    'stepsister', 'stepsisters', 'stepfather', 'stepmother', 'relative', 'relatives',
    'kin', 'kindred', 'family', 'house', 'dynasty', 'bloodline', 'stock', 'seed',
    'forebear', 'forebears', 'successor', 'successors', 'inheritor', 'inheritors',
    'offspring of', 'issue of', 'spawn', 'get', 'posterity', 'generation', 'generations',
    'patriarch', 'matriarch', 'elder', 'elders', 'younger', 'youngers', 'consort',
    'spouse', 'betrothed', 'fiancé', 'fiancée', 'ward', 'guardian', 'mentor', 'disciple',
    'teacher', 'pupil', 'master', 'student', 'ally', 'allies', 'friend', 'friends',
    'companion', 'companions', 'comrade', 'comrades', 'foe', 'foes', 'enemy', 'enemies',
    'rival', 'rivals', 'chieftain', 'leader', 'follower', 'followers', 'vassal', 'vassals',
    'lord', 'liege', 'retainer', 'retainers', 'patron', 'client', 'kinswoman', 'foremother',
    'clan leader', 'tribal chief', 'house head', 'lineal', 'collateral', 'agnate', 'cognate'
]

gender_map = {
    'Arjuna': 'male', 'Bhima': 'male', 'Yudhishthira': 'male', 'Nakula': 'male',
    'Sahadeva': 'male', 'Krishna': 'male', 'Duryodhana': 'male', 'Bhishma': 'male',
    'Karna': 'male', 'Drona': 'male', 'Dhritarashtra': 'male', 'Sanjaya': 'male',
    'Abhimanyu': 'male', 'Drupada': 'male', 'Virata': 'male', 'Satyaki': 'male',
    'Pandu': 'male', 'Vyasa': 'male', 'Brahma': 'male', 'Manu': 'male',
    'Ikshvaku': 'male', 'Vivasvan': 'male', 'Narada': 'male', 'Kapila': 'male',
    'Bhrigu': 'male', 'Usanas': 'male', 'Chekitana': 'male', 'Dhrishtaketu': 'male',
    'Yodhamanyu': 'male', 'Kripa': 'male', 'Asvatthama': 'male', 'Bhurisrava': 'male',
    'Somadatta': 'male', 'Kunti': 'female', 'Subhadra': 'female', 'Draupadi': 'female',
    'Devaki': 'female', 'Sikhandi': 'male',
    'Dhrishtadyumna': 'male', 'Jayadratha': 'male', 'Shalya': 'male', 'Ghatotkacha': 'male',
    'Kritavarma': 'male', 'Markandeya': 'male', 'Angira': 'male', 'Vashishtha': 'male',
    'Gautama': 'male', 'Kashyapa': 'male', 'Chyavana': 'male', 'Vishvamitra': 'male',
    'Shiva': 'male', 'Indra': 'male', 'Yama': 'male', 'Mrityu': 'male', 'Agni': 'male',
    'Varuna': 'male', 'Vayu': 'male', 'Aditya': 'male', 'Surya': 'male', 'Chandra': 'male',
    'Kubera': 'male', 'Hanuman': 'male', 'Garuda': 'male', 'Vasuki': 'male', 'Shesha': 'male',
    'Alambusha': 'male', 'Bahlika': 'male', 'Shrutayudha': 'male', 'Sudakshina': 'male',
    'Sarvaga': 'male', 'Pradyumna': 'male', 'Samba': 'male', 'Satyavati': 'female',
    'Amba': 'female', 'Ambika': 'female', 'Ambalika': 'female',
    'Gandhari': 'female', 'Madri': 'female', 'Shakuni': 'male', 'Balarama': 'male',
    'Parikshit': 'male', 'Janamejaya': 'male', 'Hidimbi': 'female', 'Ulupi': 'female',
    'Chitrangada': 'female', 'Devika': 'female', 'Karenumati': 'female', 'Vijaya': 'female',
    'Bhanumati': 'female', 'Vrushali': 'female', 'Ganga': 'female', 'Shantanu': 'male',
    'Vidura': 'male', 'Dushala': 'female', 'Yuyutsu': 'male', 'Laxman': 'male',
    'Chitrasena': 'male', 'Chitravarma': 'male', 'Yaudheya': 'male', 'Niramitra': 'male',
    'Suhotra': 'male', 'Iravan': 'male', 'Babhruvahana': 'male', 'Vrishasena': 'male',
    'Sushena': 'male', 'Uttara': 'female',
    'Shrutakarma': 'male', 'Sutasoma': 'male', 'Shrutasena': 'male', 'Shrutakirti': 'male',
    'Prativindhya': 'male', 'Kshatradeva': 'male', 'Kshatradharman': 'male', 'Satyajit': 'male',
    'Kekaya': 'male', 'Brihadkshatra': 'male', 'Jarasandha': 'male', 'Rukmini': 'female',
    'Satyabhama': 'female', 'Jambavati': 'female', 'Rohini': 'female', 'Aniruddha': 'male',
    'Charudeshna': 'male', 'Saurin': 'male', 'Sini': 'male', 'Vinda': 'male', 'Anuvinda': 'male',
    'Shishupala': 'male', 'Rukmi': 'male', 'Kamsa': 'male', 'Ekalavya': 'male', 'Srinjaya': 'male',
    'Dhruva': 'male', 'Ugrasena': 'male', 'Vasudeva (father)': 'male', 'Rohini (mother)': 'female'
}

# Precompute lowercase sets for faster lookups
person_names_set = set(name.lower() for name in person_names)
locations_set = set(loc.lower() for loc in locations)
relation_keywords_set = set(rel.lower() for rel in relation_keywords)

# Normalize person names and gender map
name_to_canonical = {k.lower(): v.lower() for k, v in entity_map.items()}
normalized_person_names = []
for name in person_names:
    name_lower = name.lower()
    if name_lower in name_to_canonical:
        normalized_person_names.append(name_to_canonical[name_lower])
    else:
        normalized_person_names.append(name_lower)
person_names = list(dict.fromkeys(normalized_person_names))  # Remove duplicates
person_names_set = set(person_names)  # Update the set with normalized names

gender_map = {k.lower(): v for k, v in gender_map.items()}
for k in list(gender_map.keys()):
    if k in name_to_canonical:
        gender_map[name_to_canonical[k]] = gender_map.pop(k)

# Step 6: Define predefined relations (for Code 2)
predefined_relations = [
    # Parent-Child Relations
    ('Pandu', 'father', 'Arjuna'), ('Pandu', 'father', 'Bhima'), ('Pandu', 'father', 'Yudhishthira'),
    ('Pandu', 'father', 'Nakula'), ('Pandu', 'father', 'Sahadeva'),
    ('Kunti', 'mother', 'Arjuna'), ('Kunti', 'mother', 'Bhima'), ('Kunti', 'mother', 'Yudhishthira'),
    ('Kunti', 'mother', 'Nakula'), ('Kunti', 'mother', 'Sahadeva'),
    ('Kunti', 'mother', 'Karna'), ('Karna', 'son', 'Kunti'),
    ('Dhritarashtra', 'father', 'Duryodhana'), ('Dhritarashtra', 'father', 'Dushasana'),
    ('Dhritarashtra', 'father', 'Vikarna'), ('Gandhari', 'mother', 'Duryodhana'),
    ('Gandhari', 'mother', 'Dushasana'), ('Gandhari', 'mother', 'Vikarna'),
    ('Drupada', 'father', 'Draupadi'), ('Draupadi', 'daughter', 'Drupada'),
    ('Drupada', 'father', 'Dhrishtadyumna'), ('Dhrishtadyumna', 'son', 'Drupada'),
    ('Vasudeva', 'father', 'Krishna'), ('Krishna', 'son', 'Vasudeva'),
    ('Vasudeva', 'father', 'Subhadra'), ('Subhadra', 'daughter', 'Vasudeva'),
    ('Devaki', 'mother', 'Krishna'), ('Krishna', 'son', 'Devaki'),
    ('Devaki', 'mother', 'Subhadra'), ('Subhadra', 'daughter', 'Devaki'),
    ('Somadatta', 'father', 'Bhurisrava'), ('Bhurisrava', 'son', 'Somadatta'),
    ('Drona', 'father', 'Ashwatthama'), ('Ashwatthama', 'son', 'Drona'),
    ('Ganga', 'mother', 'Bhishma'), ('Shantanu', 'father', 'Bhishma'),
    ('Shantanu', 'father', 'Chitrangada'), ('Shantanu', 'father', 'Vichitravirya'),
    ('Vyasa', 'father', 'Dhritarashtra'), ('Vyasa', 'father', 'Pandu'), ('Vyasa', 'father', 'Vidura'),
    ('Satyavati', 'mother', 'Vyasa'),
    ('Arjuna', 'father', 'Abhimanyu'), ('Subhadra', 'mother', 'Abhimanyu'),
    ('Bhima', 'father', 'Ghatotkacha'), ('Hidimbi', 'mother', 'Ghatotkacha'),
    ('Yudhishthira', 'father', 'Yaudheya'), ('Devika', 'mother', 'Yaudheya'),
    ('Nakula', 'father', 'Niramitra'), ('Karenumati', 'mother', 'Niramitra'),
    ('Sahadeva', 'father', 'Suhotra'), ('Vijaya', 'mother', 'Suhotra'),
    ('Karna', 'father', 'Vrishasena'), ('Karna', 'father', 'Sushena'),
    ('Arjuna', 'father', 'Iravan'), ('Ulupi', 'mother', 'Iravan'),
    ('Arjuna', 'father', 'Babhruvahana'), ('Chitrangada', 'mother', 'Babhruvahana'),
    ('Abhimanyu', 'father', 'Parikshit'), ('Uttara', 'mother', 'Parikshit'),
    ('Duryodhana', 'father', 'Laxman'), ('Bhanumati', 'mother', 'Laxman'),
    ('Dushasana', 'father', 'Chitrasena'),
    ('Vikarna', 'father', 'Chitravarma'),

    # Sibling Relations
    ('Arjuna', 'brother', 'Bhima'), ('Bhima', 'brother', 'Arjuna'),
    ('Arjuna', 'brother', 'Yudhishthira'), ('Yudhishthira', 'brother', 'Arjuna'),
    ('Arjuna', 'brother', 'Nakula'), ('Nakula', 'brother', 'Arjuna'),
    ('Arjuna', 'brother', 'Sahadeva'), ('Sahadeva', 'brother', 'Arjuna'),
    ('Bhima', 'brother', 'Yudhishthira'), ('Yudhishthira', 'brother', 'Bhima'),
    ('Bhima', 'brother', 'Nakula'), ('Nakula', 'brother', 'Bhima'),
    ('Bhima', 'brother', 'Sahadeva'), ('Sahadeva', 'brother', 'Bhima'),
    ('Yudhishthira', 'brother', 'Nakula'), ('Nakula', 'brother', 'Yudhishthira'),
    ('Yudhishthira', 'brother', 'Sahadeva'), ('Sahadeva', 'brother', 'Yudhishthira'),
    ('Nakula', 'brother', 'Sahadeva'), ('Sahadeva', 'brother', 'Nakula'),
    ('Duryodhana', 'brother', 'Dushasana'), ('Dushasana', 'brother', 'Duryodhana'),
    ('Duryodhana', 'brother', 'Vikarna'), ('Vikarna', 'brother', 'Duryodhana'),
    ('Dushasana', 'brother', 'Vikarna'), ('Vikarna', 'brother', 'Dushasana'),
    ('Krishna', 'brother', 'Subhadra'), ('Subhadra', 'sister', 'Krishna'),
    ('Draupadi', 'sister', 'Dhrishtadyumna'), ('Dhrishtadyumna', 'brother', 'Draupadi'),  # Added
    ('Dhritarashtra', 'brother', 'Pandu'), ('Pandu', 'brother', 'Dhritarashtra'),
    ('Dhritarashtra', 'brother', 'Vidura'), ('Vidura', 'brother', 'Dhritarashtra'),
    ('Pandu', 'brother', 'Vidura'), ('Vidura', 'brother', 'Pandu'),

    # Spouse Relations
    ('Arjuna', 'husband', 'Subhadra'), ('Subhadra', 'wife', 'Arjuna'),
    ('Arjuna', 'husband', 'Draupadi'), ('Draupadi', 'wife', 'Arjuna'),
    ('Arjuna', 'husband', 'Ulupi'), ('Ulupi', 'wife', 'Arjuna'),
    ('Arjuna', 'husband', 'Chitrangada'), ('Chitrangada', 'wife', 'Arjuna'),
    ('Bhima', 'husband', 'Hidimbi'), ('Hidimbi', 'wife', 'Bhima'),
    ('Bhima', 'husband', 'Draupadi'), ('Draupadi', 'wife', 'Bhima'),
    ('Yudhishthira', 'husband', 'Draupadi'), ('Draupadi', 'wife', 'Yudhishthira'),
    ('Yudhishthira', 'husband', 'Devika'), ('Devika', 'wife', 'Yudhishthira'),
    ('Nakula', 'husband', 'Draupadi'), ('Draupadi', 'wife', 'Nakula'),
    ('Nakula', 'husband', 'Karenumati'), ('Karenumati', 'wife', 'Nakula'),
    ('Sahadeva', 'husband', 'Draupadi'), ('Draupadi', 'wife', 'Sahadeva'),
    ('Sahadeva', 'husband', 'Vijaya'), ('Vijaya', 'wife', 'Sahadeva'),
    ('Duryodhana', 'husband', 'Bhanumati'), ('Bhanumati', 'wife', 'Duryodhana'),
    ('Abhimanyu', 'husband', 'Uttara'), ('Uttara', 'wife', 'Abhimanyu'),
    ('Karna', 'husband', 'Vrushali'), ('Vrushali', 'wife', 'Karna'),

    # Grandparent Relations
    ('Arjuna', 'grandfather', 'Parikshit'), ('Subhadra', 'grandmother', 'Parikshit'),
    ('Bhima', 'grandfather', 'Sutasoma'), ('Draupadi', 'grandmother', 'Sutasoma'),
    ('Yudhishthira', 'grandfather', 'Prativindhya'), ('Draupadi', 'grandmother', 'Prativindhya'),
    ('Duryodhana', 'grandfather', 'Laxman'), ('Bhanumati', 'grandmother', 'Laxman'),
    ('Dushasana', 'grandfather', 'Chitrasena'),
    ('Bhishma', 'grandfather', 'Duryodhana'), ('Bhishma', 'grandfather', 'Arjuna'),
    ('Shantanu', 'grandfather', 'Dhritarashtra'), ('Shantanu', 'grandfather', 'Pandu'),

    # Children of Draupadi (with all Pandavas)
    ('Draupadi', 'mother', 'Prativindhya'), ('Yudhishthira', 'father', 'Prativindhya'),
    ('Draupadi', 'mother', 'Sutasoma'), ('Bhima', 'father', 'Sutasoma'),
    ('Draupadi', 'mother', 'Shrutakarma'), ('Arjuna', 'father', 'Shrutakarma'),
    ('Draupadi', 'mother', 'Shrutasena'), ('Nakula', 'father', 'Shrutasena'),
    ('Draupadi', 'mother', 'Shrutakirti'), ('Sahadeva', 'father', 'Shrutakirti'),

    # Cousin Relations
    ('Arjuna', 'cousin', 'Krishna'), ('Krishna', 'cousin', 'Arjuna'),
    ('Bhima', 'cousin', 'Krishna'), ('Krishna', 'cousin', 'Bhima'),
    ('Yudhishthira', 'cousin', 'Krishna'), ('Krishna', 'cousin', 'Yudhishthira'),
    ('Nakula', 'cousin', 'Krishna'), ('Krishna', 'cousin', 'Nakula'),
    ('Sahadeva', 'cousin', 'Krishna'), ('Krishna', 'cousin', 'Sahadeva'),
    ('Arjuna', 'cousin', 'Balarama'), ('Balarama', 'cousin', 'Arjuna'),
    ('Bhima', 'cousin', 'Balarama'), ('Balarama', 'cousin', 'Bhima'),
    # In-Law Relations
    ('Krishna', 'brother-in-law', 'Arjuna'), ('Arjuna', 'brother-in-law', 'Krishna'),
    ('Krishna', 'brother-in-law', 'Bhima'), ('Bhima', 'brother-in-law', 'Krishna'),
    ('Krishna', 'brother-in-law', 'Yudhishthira'), ('Yudhishthira', 'brother-in-law', 'Krishna'),
    ('Krishna', 'brother-in-law', 'Nakula'), ('Nakula', 'brother-in-law', 'Krishna'),
    ('Krishna', 'brother-in-law', 'Sahadeva'), ('Sahadeva', 'brother-in-law', 'Krishna'),
    ('Pandu', 'father-in-law', 'Draupadi'), ('Draupadi', 'daughter-in-law', 'Pandu'),
    ('Kunti', 'mother-in-law', 'Draupadi'), ('Draupadi', 'daughter-in-law', 'Kunti'),
    ('Dhritarashtra', 'father-in-law', 'Draupadi'), ('Draupadi', 'daughter-in-law', 'Dhritarashtra'),
    ('Gandhari', 'mother-in-law', 'Draupadi'), ('Draupadi', 'daughter-in-law', 'Gandhari'),
    ('Drupada', 'father-in-law', 'Yudhishthira'), ('Yudhishthira', 'son-in-law', 'Drupada'),
    ('Vasudeva', 'father-in-law', 'Arjuna'), ('Arjuna', 'son-in-law', 'Vasudeva'),
    ('Devaki', 'mother-in-law', 'Arjuna'), ('Arjuna', 'daughter-in-law', 'Devaki'),

    # Uncle/Aunt and Nephew/Niece Relations
    ('Shakuni', 'uncle', 'Duryodhana'), ('Duryodhana', 'nephew', 'Shakuni'),
    ('Shakuni', 'uncle', 'Dushasana'), ('Dushasana', 'nephew', 'Shakuni'),
    ('Vikarna', 'uncle', 'Laxman'), ('Laxman', 'nephew', 'Vikarna'),
    ('Satyavati', 'aunt', 'Vyasa'), ('Vyasa', 'nephew', 'Satyavati'),
    ('Kunti', 'aunt', 'Abhimanyu'), ('Abhimanyu', 'nephew', 'Kunti'),
    ('Vidura', 'uncle', 'Arjuna'), ('Arjuna', 'nephew', 'Vidura'),

    # Teacher-Disciple Relations
    ('Drona', 'teacher', 'Arjuna'), ('Drona', 'teacher', 'Bhima'),
    ('Drona', 'teacher', 'Yudhishthira'), ('Drona', 'teacher', 'Nakula'),
    ('Drona', 'teacher', 'Sahadeva'), ('Drona', 'teacher', 'Ashwatthama'),
    ('Drona', 'teacher', 'Duryodhana'), ('Drona', 'teacher', 'Dushasana'),
    ('Kripa', 'teacher', 'Ashwatthama'), ('Kripa', 'teacher', 'Arjuna'),
    ('Kripa', 'teacher', 'Duryodhana'), ('Duryodhana', 'disciple', 'Kripa'),
    ('bhishma', 'teacher', 'arjuna'),('bhishma', 'teacher', 'yudhishthira'),
    ('bhishma', 'teacher', 'bhima'),('bhishma', 'teacher', 'nakula'),
    ('bhishma', 'teacher', 'sahadeva'),

    # Friend and Ally Relations
    ('Karna', 'friend', 'Duryodhana'), ('Duryodhana', 'friend', 'Karna'),
    ('Ashwatthama', 'ally', 'Duryodhana'), ('Duryodhana', 'ally', 'Ashwatthama'),
    ('Krishna', 'friend', 'Arjuna'), ('Arjuna', 'friend', 'Krishna'),
    ('Bhishma', 'guardian', 'Duryodhana'), ('Duryodhana', 'ward', 'Bhishma'),
    ('Bhishma', 'guardian', 'Arjuna'), ('Arjuna', 'ward', 'Bhishma'),


    # Other Relations
    ('Arjuna', 'family', 'Krishna'),('Duryodhana', 'family', 'Dhritarashtra'),
    ('krishna', 'friend', 'arjuna'),('arjuna', 'friend', 'krishna'),
    ('duryodhana', 'friend', 'karna'),('karna', 'friend', 'duryodhana'),
    ('krishna', 'friend', 'yudhishthira'),('subhadra', 'sister', 'krishna'),
    ('draupadi', 'sister', 'dhrishtadyumna'),('dushala', 'sister', 'duryodhana'),
    ('drona', 'teacher', 'arjuna'),('drona', 'teacher', 'yudhishthira'),
    ('drona', 'teacher', 'bhima'),('kripa', 'teacher', 'duryodhana'),
    ('kripa', 'teacher', 'arjuna'),('bhishma', 'teacher', 'pandavas'),
    ('arjuna', 'ally', 'krishna'),('duryodhana', 'ally', 'karna'),
    ('duryodhana', 'ally', 'ashwatthama'),('yudhishthira', 'ally', 'krishna')

]
single_person_mappings = {
    'son': {
        'Arjuna': ('Kunti', 'mother'), 'Bhima': ('Kunti', 'mother'),
        'Yudhishthira': ('Kunti', 'mother'), 'Nakula': ('Kunti', 'mother'),
        'Sahadeva': ('Kunti', 'mother'), 'Krishna': ('Devaki', 'mother'),
        'Duryodhana': ('Dhritarashtra', 'father'), 'Bhurisrava': ('Somadatta', 'father'),
        'Ashwatthama': ('Drona', 'father'), 'Abhimanyu': ('Arjuna', 'father'),
        'Karna': ('Kunti', 'mother'), 'Partha': ('Kunti', 'mother'),
        'Dhrishtadyumna': ('Drupada', 'father'),
        'Ghatotkacha': ('Bhima', 'father'),
        'Yaudheya': ('Yudhishthira', 'father'),
        'Niramitra': ('Nakula', 'father'),
        'Suhotra': ('Sahadeva', 'father'),
        'Vrishasena': ('Karna', 'father'),
        'Sushena': ('Karna', 'father'),
        'Iravan': ('Arjuna', 'father'),
        'Babhruvahana': ('Arjuna', 'father'),
        'Parikshit': ('Abhimanyu', 'father'),
        'Laxman': ('Duryodhana', 'father'),
        'Chitrasena': ('Dushasana', 'father'),
        'Chitravarma': ('Vikarna', 'father'),
        'Dhritarashtra': ('Vyasa', 'father'),
        'Pandu': ('Vyasa', 'father'),
        'Vidura': ('Vyasa', 'father'),
        'Bhishma': ('Shantanu', 'father'),
    },
    'daughter': {
        'Draupadi': ('Drupada', 'father'), 'Subhadra': ('Vasudeva', 'father'),
        'Uttara': ('Virata', 'father'),
    },
    'father': {
        'Drupada': ('Draupadi', 'daughter'), 'Dhritarashtra': ('Duryodhana', 'son'),
        'Pandu': ('Arjuna', 'son'), 'Vasudeva': ('Krishna', 'son'),
        'Somadatta': ('Bhurisrava', 'son'), 'Drona': ('Ashwatthama', 'son'),
        'Pandu': ('Yudhishthira', 'son'), 'Pandu': ('Bhima', 'son'),
        'Drupada': ('Dhrishtadyumna', 'son'), 'Vasudeva': ('Subhadra', 'daughter'),
        'Arjuna': ('Abhimanyu', 'son'),
        'Bhima': ('Ghatotkacha', 'son'),
        'Yudhishthira': ('Yaudheya', 'son'),
        'Nakula': ('Niramitra', 'son'),
        'Sahadeva': ('Suhotra', 'son'),
        'Karna': ('Vrishasena', 'son'),
        'Arjuna': ('Iravan', 'son'),
        'Arjuna': ('Babhruvahana', 'son'),
        'Abhimanyu': ('Parikshit', 'son'),
        'Duryodhana': ('Laxman', 'son'),
        'Dushasana': ('Chitrasena', 'son'),
        'Vikarna': ('Chitravarma', 'son'),
        'Vyasa': ('Dhritarashtra', 'son'),
        'Shantanu': ('Bhishma', 'son'),
        'Virata': ('Uttara', 'daughter'),
    },
    'mother': {
        'Kunti': ('Arjuna', 'son'), 'Devaki': ('Krishna', 'son'),
        'Pritha': ('Arjuna', 'son'), 'Kunti': ('Yudhishthira', 'son'),
        'Kunti': ('Karna', 'son'), 'Pritha': ('Karna', 'son'),
        'Pritha': ('Yudhishthira', 'son'), 'Devaki': ('Subhadra', 'daughter'),
        'Gandhari': ('Duryodhana', 'son'),
        'Subhadra': ('Abhimanyu', 'son'),
        'Hidimbi': ('Ghatotkacha', 'son'),
        'Devika': ('Yaudheya', 'son'),
        'Karenumati': ('Niramitra', 'son'),
        'Vijaya': ('Suhotra', 'son'),
        'Ulupi': ('Iravan', 'son'),
        'Chitrangada': ('Babhruvahana', 'son'),
        'Uttara': ('Parikshit', 'son'),
        'Bhanumati': ('Laxman', 'son'),
        'Vrushali': ('Vrishasena', 'son'),
        'Ganga': ('Bhishma', 'son'),
        'Satyavati': ('Vyasa', 'son'),
    },
    'brother': {
        'Arjuna': ('Bhima', 'brother'), 'Bhima': ('Arjuna', 'brother'),
        'Yudhishthira': ('Arjuna', 'brother'), 'Nakula': ('Sahadeva', 'brother'),
        'Sahadeva': ('Nakula', 'brother'), 'Arjuna': ('Yudhishthira', 'brother'),
        'Bhima': ('Yudhishthira', 'brother'), 'Yudhishthira': ('Bhima', 'brother'),
        'Arjuna': ('Nakula', 'brother'), 'Arjuna': ('Sahadeva', 'brother'),
        'Bhima': ('Nakula', 'brother'), 'Bhima': ('Sahadeva', 'brother'),
        'Krishna': ('Subhadra', 'sister'),
        'Duryodhana': ('Dushasana', 'brother'),
        'Dushasana': ('Duryodhana', 'brother'),
        'Vikarna': ('Duryodhana', 'brother'),
        'Dhrishtadyumna': ('Draupadi', 'sister'),
        'Dhritarashtra': ('Pandu', 'brother'),
        'Pandu': ('Vidura', 'brother'),
        'Krishna': ('Balarama', 'brother'),
    },
    'sister': {
        'Subhadra': ('Krishna', 'brother'),
        'Draupadi': ('Dhrishtadyumna', 'brother'),
        'Dushala': ('Duryodhana', 'brother'),
        'Dushasana': ('Duryodhana', 'brother'),
        'Vikarna': ('Duryodhana', 'brother'),
        'Dhrishtadyumna': ('Draupadi', 'sister'),
    },
    'husband': {
        'Arjuna': ('Subhadra', 'wife'), 'Arjuna': ('Draupadi', 'wife'),
        'Yudhishthira': ('Draupadi', 'wife'),
        'Bhima': ('Draupadi', 'wife'),
        'Nakula': ('Draupadi', 'wife'),
        'Sahadeva': ('Draupadi', 'wife'),
        'Bhima': ('Hidimbi', 'wife'),
        'Arjuna': ('Ulupi', 'wife'),
        'Arjuna': ('Chitrangada', 'wife'),
        'Yudhishthira': ('Devika', 'wife'),
        'Nakula': ('Karenumati', 'wife'),
        'Sahadeva': ('Vijaya', 'wife'),
        'Duryodhana': ('Bhanumati', 'wife'),
        'Karna': ('Vrushali', 'wife'),
        'Abhimanyu': ('Uttara', 'wife'),
    },
    'wife': {
        'Subhadra': ('Arjuna', 'husband'), 'Draupadi': ('Arjuna', 'husband'),
        'Draupadi': ('Yudhishthira', 'husband'),
        'Draupadi': ('Bhima', 'husband'),
        'Draupadi': ('Nakula', 'husband'),
        'Draupadi': ('Sahadeva', 'husband'),
        'Hidimbi': ('Bhima', 'husband'),
        'Ulupi': ('Arjuna', 'husband'),
        'Chitrangada': ('Arjuna', 'husband'),
        'Devika': ('Yudhishthira', 'husband'),
        'Karenumati': ('Nakula', 'husband'),
        'Vijaya': ('Sahadeva', 'husband'),
        'Bhanumati': ('Duryodhana', 'husband'),
        'Vrushali': ('Karna', 'husband'),
        'Uttara': ('Abhimanyu', 'husband'),
    },
    'brother-in-law': {
        'Krishna': ('Arjuna', 'brother-in-law'), 'Arjuna': ('Krishna', 'brother-in-law'),
        'Krishna': ('Bhima', 'brother-in-law'), 'Bhima': ('Krishna', 'brother-in-law'),
        'Krishna': ('Yudhishthira', 'brother-in-law'),
        'Yudhishthira': ('Krishna', 'brother-in-law'),
        'Krishna': ('Nakula', 'brother-in-law'),
        'Nakula': ('Krishna', 'brother-in-law'),
        'Krishna': ('Sahadeva', 'brother-in-law'),
        'Sahadeva': ('Krishna', 'brother-in-law'),
        'Dhrishtadyumna': ('Yudhishthira', 'brother-in-law'),
        'Yudhishthira': ('Dhrishtadyumna', 'brother-in-law'),
    },
    'sister-in-law': {
        'Subhadra': ('Yudhishthira', 'brother-in-law'),
        'Draupadi': ('Subhadra', 'sister-in-law'),
        'Subhadra': ('Draupadi', 'sister-in-law'),
        'Bhanumati': ('Draupadi', 'sister-in-law'),
    },
    'cousin': {
        'Krishna': ('Arjuna', 'cousin'), 'Arjuna': ('Krishna', 'cousin'),
        'Krishna': ('Bhima', 'cousin'), 'Bhima': ('Krishna', 'cousin'),
        'Krishna': ('Yudhishthira', 'cousin'),
        'Yudhishthira': ('Krishna', 'cousin'),
        'Krishna': ('Nakula', 'cousin'),
        'Nakula': ('Krishna', 'cousin'),
        'Krishna': ('Sahadeva', 'cousin'),
        'Sahadeva': ('Krishna', 'cousin'),
        'Balarama': ('Arjuna', 'cousin'),
        'Arjuna': ('Balarama', 'cousin'),
    },
    'grandfather': {
        'Arjuna': ('Bhishma', 'grandfather'), 'Duryodhana': ('Bhishma', 'grandfather'),
        'Arjuna': ('Parikshit', 'grandson'),
        'Bhima': ('Sutasoma', 'grandson'),
        'Yudhishthira': ('Prativindhya', 'grandson'),
        'Duryodhana': ('Laxman', 'grandson'),
        'Shantanu': ('Dhritarashtra', 'grandson'),
    },
    'grandmother': {
        'Kunti': ('Abhimanyu', 'grandson'),
        'Gandhari': ('Laxman', 'grandson'),
        'Subhadra': ('Parikshit', 'grandson'),
        'Draupadi': ('Prativindhya', 'grandson'),
        'Bhanumati': ('Laxman', 'grandson'),
    },
    'grandson': {
        'Abhimanyu': ('Bhishma', 'grandfather'),
        'Parikshit': ('Arjuna', 'grandfather'),
        'Sutasoma': ('Bhima', 'grandfather'),
        'Prativindhya': ('Yudhishthira', 'grandfather'),
        'Laxman': ('Duryodhana', 'grandfather'),
        'Dhritarashtra': ('Shantanu', 'grandfather'),
    },
    'uncle': {
        'Shakuni': ('Duryodhana', 'nephew'),
        'Vidura': ('Arjuna', 'nephew'),
        'Krishna': ('Abhimanyu', 'nephew'),
    },
    'aunt': {
        'Kunti': ('Abhimanyu', 'nephew'),
        'Gandhari': ('Arjuna', 'nephew'),
    },
    'nephew': {
        'Duryodhana': ('Shakuni', 'uncle'),
        'Arjuna': ('Vidura', 'uncle'),
        'Abhimanyu': ('Krishna', 'uncle'),
    },
    'niece': {
        'Uttara': ('Krishna', 'brother-in-law'),
    },
    'father-in-law': {
        'Drupada': ('Yudhishthira', 'son-in-law'),
        'Vasudeva': ('Arjuna', 'son-in-law'),
        'Virata': ('Abhimanyu', 'son-in-law'),
    },
    'mother-in-law': {
        'Kunti': ('Draupadi', 'daughter-in-law'),
        'Gandhari': ('Draupadi', 'daughter-in-law'),
        'Devaki': ('Arjuna', 'son-in-law'),
    },
    'son-in-law': {
        'Yudhishthira': ('Drupada', 'father-in-law'),
        'Arjuna': ('Vasudeva', 'father-in-law'),
        'Abhimanyu': ('Virata', 'father-in-law'),
    },
    'daughter-in-law': {
        'Draupadi': ('Kunti', 'mother-in-law'),
        'Subhadra': ('Kunti', 'mother-in-law'),
    },
    'friend': {
        'Arjuna': ('Krishna', 'friend'),
        'Karna': ('Duryodhana', 'friend'),
    },
    'teacher': {
        'Drona': ('Arjuna', 'disciple'),
        'Drona': ('Ashwatthama', 'disciple'),
        'Kripa': ('Duryodhana', 'disciple'),
    },
    'disciple': {
        'Arjuna': ('Drona', 'teacher'),
        'Ashwatthama': ('Drona', 'teacher'),
        'Duryodhana': ('Kripa', 'teacher'),
    }
}

# Expanded invalid_pairs to include incorrect Krishna-Bhima brother relation
invalid_pairs = {
    ('Brahma', 'son', 'Kunti'), ('Brahma', 'daughter', 'Kunti'),
    ('Bhishma', 'son', 'Pritha'), ('Bhishma', 'daughter', 'Pritha'),
    ('Yodhamanyu', 'son', 'Subhadra'), ('Yodhamanyu', 'son', 'Draupadi'),
    ('Kunti', 'son', 'Unknown'), ('Kunti', 'son', 'Brahma'),
    ('Draupadi', 'son', 'Unknown'), ('Subhadra', 'son', 'Unknown'),
    ('Dhritarashtra', 'son', 'Pandu'), ('Madhava', 'son', 'Pandu'),
    ('Drupada', 'son', 'Subhadra'), ('Drupada', 'daughter', 'Subhadra'),
    ('Bhishma', 'son', 'Unknown'), ('Brahma', 'son', 'Unknown'),
    ('Bhishma', 'son', 'Somadatta'), ('Pandu', 'son', 'Drupada'),
    ('Madhava', 'son', 'Unknown'), ('Pritha', 'son', 'Unknown'),
    ('Arjuna', 'son', 'Pandu'), ('Yodhamanyu', 'son', 'Unknown'),
    ('Dhritarashtra', 'son', 'Unknown'), ('Pandu', 'son', 'Unknown'),
    ('Arjuna', 'father', 'Unknown'), ('Drupada', 'son', 'Draupadi'),
    ('Draupadi', 'mother', 'Drupada'), ('Krishna', 'son', 'Pandu'),
    ('Bhishma', 'grandsire', 'Unknown'), ('Drupada', 'son', 'Unknown'),
    ('Arjuna', 'maternal uncle', 'Unknown'), ('Arjuna', 'grandson', 'Unknown'),
    ('Kunti', 'father', 'Unknown'), ('Vasuki', 'father', 'Unknown'),
    ('Vayu', 'grandfather', 'Unknown'), ('Krishna', 'father', 'Draupadi'),
    ('Krishna', 'grandfather', 'Draupadi'), ('Krishna', 'mother', 'Draupadi'),
    ('Krishna', 'grandmother', 'Draupadi'),
    ('Krishna', 'brother', 'Bhima'), ('Bhima', 'brother', 'Krishna'),
    ('Arjuna', 'father', 'Kunti'), ('Kunti', 'daughter', 'Arjuna'),
    ('Draupadi', 'father', 'Yudhishthira'), ('Yudhishthira', 'mother', 'Draupadi'),
    ('Bhishma', 'father', 'Arjuna'), ('Bhishma', 'husband', 'Gandhari'),
    ('Duryodhana', 'brother', 'Arjuna'), ('Arjuna', 'brother', 'Duryodhana'),
    ('Krishna', 'son', 'Arjuna'), ('Arjuna', 'father', 'Krishna'),
    ('Subhadra', 'mother', 'Krishna'), ('Krishna', 'father', 'Subhadra'),
    ('Draupadi', 'sister', 'Arjuna'), ('Arjuna', 'brother', 'Draupadi'),
    ('Karna', 'father', 'Kunti'), ('Kunti', 'daughter', 'Karna'),
    ('Abhimanyu', 'father', 'Arjuna'), ('Arjuna', 'son', 'Abhimanyu'),
    ('Parikshit', 'father', 'Abhimanyu'), ('Abhimanyu', 'grandfather', 'Parikshit'),
    ('Gandhari', 'daughter', 'Duryodhana'), ('Duryodhana', 'father', 'Gandhari'),
    ('Shakuni', 'father', 'Duryodhana'), ('Duryodhana', 'nephew', 'Kunti'),
    ('Balarama', 'brother', 'Arjuna'), ('Arjuna', 'brother', 'Balarama'),
    ('Drona', 'father', 'Arjuna'), ('Arjuna', 'son', 'Drona'),
    ('Ashwatthama', 'brother', 'Arjuna'), ('Draupadi', 'mother', 'Krishna'),
    ('Krishna', 'husband', 'Draupadi'), ('Subhadra', 'wife', 'Yudhishthira'),
    ('Uttara', 'mother', 'Arjuna'), ('Arjuna', 'son', 'Uttara'),
    ('Virata', 'father', 'Arjuna'), ('Arjuna', 'son', 'Virata'),
    ('Dushala', 'brother', 'Arjuna'), ('Arjuna', 'sister', 'Dushala'),
    ('Vyasa', 'son', 'Dhritarashtra'), ('Dhritarashtra', 'father', 'Vyasa'),
    ('Shantanu', 'son', 'Bhishma'), ('Bhishma', 'father', 'Shantanu'),
    ('Ganga', 'daughter', 'Bhishma'), ('Bhishma', 'father', 'Ganga'),
    ('Satyavati', 'daughter', 'Vyasa'), ('Vyasa', 'father', 'Satyavati'),
    ('Dhrishtadyumna', 'husband', 'Draupadi'), ('Draupadi', 'wife', 'Dhrishtadyumna'),
    ('Kripa', 'father', 'Arjuna'), ('Arjuna', 'son', 'Kripa'),
    ('Bhanumati', 'sister', 'Duryodhana'), ('Duryodhana', 'brother', 'Bhanumati'),
    ('Vrushali', 'sister', 'Karna'), ('Karna', 'brother', 'Vrushali'),
    ('Hidimbi', 'sister', 'Bhima'), ('Bhima', 'brother', 'Hidimbi'),
    ('Ulupi', 'sister', 'Arjuna'), ('Arjuna', 'brother', 'Ulupi'),
    ('Chitrangada', 'sister', 'Arjuna'), ('Arjuna', 'brother', 'Chitrangada'),
    ('Devika', 'sister', 'Yudhishthira'), ('Yudhishthira', 'brother', 'Devika'),
    ('Karenumati', 'sister', 'Nakula'), ('Nakula', 'brother', 'Karenumati'),
    ('Vijaya', 'sister', 'Sahadeva'), ('Sahadeva', 'brother', 'Vijaya'),
    ('Krishna', 'teacher', 'Arjuna'), ('Arjuna', 'disciple', 'Krishna'),
    ('Drona', 'brother', 'Arjuna'), ('Arjuna', 'brother', 'Drona'),
    ('Vasudeva', 'son', 'Krishna'), ('Krishna', 'father', 'Vasudeva'),
    ('Devaki', 'daughter', 'Krishna'), ('Krishna', 'father', 'Devaki'),
    ('Nakula', 'father', 'Arjuna'), ('Arjuna', 'son', 'Nakula'),
    ('Sahadeva', 'father', 'Bhima'), ('Bhima', 'son', 'Sahadeva'),
    ('Yudhishthira', 'father', 'Arjuna'), ('Arjuna', 'son', 'Yudhishthira'),
    ('Karna', 'brother', 'Duryodhana'), ('Duryodhana', 'brother', 'Karna'),
    ('Subhadra', 'sister', 'Draupadi'), ('Draupadi', 'sister', 'Subhadra'),
    ('Bhishma', 'brother', 'Pandu'), ('Pandu', 'brother', 'Bhishma'),
    ('Ghatotkacha', 'father', 'Bhima'), ('Bhima', 'son', 'Ghatotkacha'),
    ('Yaudheya', 'father', 'Yudhishthira'), ('Yudhishthira', 'son', 'Yaudheya'),
    ('Niramitra', 'father', 'Nakula'), ('Nakula', 'son', 'Niramitra'),
    ('Suhotra', 'father', 'Sahadeva'), ('Sahadeva', 'son', 'Suhotra'),
    ('Iravan', 'father', 'Arjuna'), ('Arjuna', 'son', 'Iravan'),
    ('Babhruvahana', 'father', 'Arjuna'), ('Arjuna', 'son', 'Babhruvahana'),
    ('Vrishasena', 'father', 'Karna'), ('Karna', 'son', 'Vrishasena'),
    ('Sushena', 'father', 'Karna'), ('Karna', 'son', 'Sushena'),
    ('Laxman', 'father', 'Duryodhana'), ('Duryodhana', 'son', 'Laxman'),
    ('Chitrasena', 'father', 'Dushasana'), ('Dushasana', 'son', 'Chitrasena'),
    ('Chitravarma', 'father', 'Vikarna'), ('Vikarna', 'son', 'Chitravarma'),
    ('Arjuna', 'grandfather', 'Abhimanyu'), ('Abhimanyu', 'grandson', 'Arjuna'),
    ('Bhima', 'grandfather', 'Ghatotkacha'), ('Ghatotkacha', 'grandson', 'Bhima'),
    ('Drona', 'father', 'Duryodhana'), ('Duryodhana', 'son', 'Drona'),
    ('Kripa', 'father', 'Duryodhana'), ('Duryodhana', 'son', 'Kripa'),
    ('Shalya', 'father', 'Arjuna'), ('Arjuna', 'son', 'Shalya'),
    ('Satyaki', 'brother', 'Arjuna'), ('Arjuna', 'brother', 'Satyaki'),
    ('Jayadratha', 'brother', 'Duryodhana'), ('Duryodhana', 'brother', 'Jayadratha'),
    ('Krishna', 'father', 'Abhimanyu'), ('Abhimanyu', 'son', 'Krishna'),
    ('Subhadra', 'mother', 'Parikshit'), ('Parikshit', 'son', 'Subhadra'),
    ('Uttara', 'wife', 'Arjuna'), ('Arjuna', 'husband', 'Uttara'),
    ('Bhanumati', 'mother', 'Duryodhana'), ('Duryodhana', 'father', 'Bhanumati'),
    ('Vrushali', 'mother', 'Karna'), ('Karna', 'father', 'Vrushali'),
    ('Hidimbi', 'mother', 'Bhima'), ('Bhima', 'father', 'Hidimbi'),
    ('Ulupi', 'mother', 'Arjuna'), ('Arjuna', 'father', 'Ulupi'),
    ('Chitrangada', 'mother', 'Arjuna'), ('Arjuna', 'father', 'Chitrangada'),
    ('Devika', 'mother', 'Yudhishthira'), ('Yudhishthira', 'father', 'Devika'),
    ('Karenumati', 'mother', 'Nakula'), ('Nakula', 'father', 'Karenumati'),
    ('Vijaya', 'mother', 'Sahadeva'), ('Sahadeva', 'father', 'Vijaya'),
    ('Draupadi', 'daughter', 'Yudhishthira'), ('Yudhishthira', 'father', 'Draupadi'),
    ('Subhadra', 'daughter', 'Arjuna'), ('Arjuna', 'father', 'Subhadra'),
    ('Kunti', 'wife', 'Pandu'), ('Pandu', 'husband', 'Kunti'),
    ('Gandhari', 'wife', 'Dhritarashtra'), ('Dhritarashtra', 'husband', 'Gandhari'),
    ('Devaki', 'wife', 'Vasudeva'), ('Vasudeva', 'husband', 'Devaki'),
    ('Dushala', 'wife', 'Jayadratha'), ('Jayadratha', 'husband', 'Dushala'),
    ('Arjuna', 'cousin', 'Duryodhana'), ('Duryodhana', 'cousin', 'Arjuna'),
    ('Karna', 'cousin', 'Arjuna'), ('Arjuna', 'cousin', 'Karna'),
    ('Bhishma', 'grandfather', 'Kunti'), ('Kunti', 'grandson', 'Bhishma'),
    ('Shantanu', 'father', 'Pandu'), ('Pandu', 'son', 'Shantanu'),
    ('Vyasa', 'grandfather', 'Arjuna'), ('Arjuna', 'grandson', 'Vyasa'),
    ('Drona', 'teacher', 'Krishna'), ('Krishna', 'disciple', 'Drona'),
    ('Kripa', 'teacher', 'Krishna'), ('Krishna', 'disciple', 'Kripa'),
    ('Vidura', 'father', 'Arjuna'), ('Arjuna', 'son', 'Vidura'),
    ('Vidura', 'brother', 'Arjuna'), ('Arjuna', 'brother', 'Vidura'),
    ('sahadeva', 'husband', 'arjuna'),
    ('arjuna', 'wife', 'sahadeva'),
    # From Cell 2 output
    ('dhritarashtra', 'son', 'pandu'),('sanjaya', 'teacher', 'drona'),
    ('pandu', 'son', 'drupada'),('yodhamanyu', 'son', 'subhadra'),
    ('bhishma', 'son', 'somadatta'),('krishna', 'son', 'pandu'),
    ('drupada', 'son', 'draupadi'),('drupada', 'lord', 'subhadra'),
    ('drupada', 'son', 'subhadra'),('arjuna', 'son', 'pandu'),
    ('sanjaya', 'lord', 'arjuna'),('sanjaya', 'foes', 'krishna'),
    ('brahma', 'son', 'kunti'),('sanjaya', 'lord', 'krishna'),
    ('krishna', 'lord', 'arjuna')
}

# Step 7: Optimized Helper Functions for Code 2
def clean_text(text):
    text = unicodedata.normalize('NFKD', str(text)).encode('ascii', 'ignore').decode('ascii')
    text = re.sub(r'[^\x20-\x7E]', ' ', text)
    return text.strip() or "empty"

def tokenize(text):
    tokens = []
    text = clean_text(text)
    doc = nlp(text)
    i = 0
    while i < len(doc):
        matched = False
        for j in range(3, 0, -1):  # Reduce the range to 3 to speed up
            if i + j <= len(doc):
                candidate = ' '.join(t.text for t in doc[i:i+j])
                candidate_lower = candidate.lower()
                if (candidate_lower in person_names_set or
                    candidate_lower in locations_set or
                    candidate_lower in relation_keywords_set):
                    tokens.append(candidate)
                    i += j
                    matched = True
                    break
        if not matched:
            token = doc[i]
            token_lower = token.text.lower()
            if (not token.is_stop or
                token_lower in person_names_set or
                token_lower in locations_set or
                token_lower in relation_keywords_set):
                tokens.append(token.text)
            i += 1
    return [t for t in tokens if t]

def tag_tokens(tokens):
    tagged = []
    i = 0
    while i < len(tokens):
        token = tokens[i]
        token_lower = token.lower()
        if token_lower in person_names_set:
            tagged.append((token, 'B-PER'))
        elif token_lower in locations_set:
            tagged.append((token, 'B-LOC'))
        elif token_lower in relation_keywords_set:
            tagged.append((token, 'B-REL'))
            j = i + 1
            while j < len(tokens):
                combined = ' '.join(tokens[i:j+1]).lower()
                if combined in relation_keywords_set:
                    tagged.append((tokens[j], 'I-REL'))
                    j += 1
                else:
                    break
            i = j - 1
        else:
            tagged.append((token, 'O'))
        i += 1
    return tagged


# extracting the relation Validate relations with gender rules and invalid_pairs using predefined relations and tesxt base extraction from tokens
def extract_relations(tagged_tokens, chapter, verse):
    relations = []
    untagged_keywords = []
    potential_relations = []
    predefined_family_pairs = {(r[0], r[2]) for r in predefined_relations if r[1] == 'family'}

    persons = [token.title() for token, tag in tagged_tokens if tag == 'B-PER']
    normalized_persons = [entity_map.get(p, p) for p in persons]

    for i, p1 in enumerate(persons):
        for p2 in persons[i+1:]:
            p1_norm = entity_map.get(p1, p1)
            p2_norm = entity_map.get(p2, p2)
            for rel in ['son', 'father', 'mother', 'brother', 'family', 'cousin', 'brother-in-law']:
                if (p1_norm, rel, p2_norm) in predefined_relations and (p1_norm, rel, p2_norm) not in invalid_pairs:
                    relations.append((p1_norm, rel, p2_norm))
                    reverse_rel = single_person_mappings.get(rel, {}).get(p2_norm, (None, None))[1]
                    if reverse_rel and (p2_norm, reverse_rel, p1_norm) not in invalid_pairs:
                        relations.append((p2_norm, reverse_rel, p1_norm))

    for i, (token, tag) in enumerate(tagged_tokens):
        if tag == 'B-REL':
            keyword = token.lower()
            if keyword in ['sons', 'child', 'children', 'offspring', 'son of', 'children of', 'child of', 'born', 'scion', 'issue', 'begotten']:
                keyword = 'son'
            elif keyword in ['daughters', 'daughter of']:
                keyword = 'daughter'
            elif keyword in ['fathers', 'parents', 'ancestors', 'forefather', 'progenitor', 'sire', 'ancestry', 'heritage', 'clan', 'tribe', 'household']:
                keyword = 'father'
            elif keyword in ['brothers', 'siblings', 'brethren']:
                keyword = 'brother'
            elif keyword in ['sisters']:
                keyword = 'sister'
            elif keyword in ['cousins']:
                keyword = 'cousin'
            elif keyword in ['grandfathers', 'grandsons', 'maternal uncles', 'descendants']:
                keyword = keyword.rstrip('s')
            elif keyword in ['kin', 'kinsmen', 'relatives', 'born to', 'born of', 'relative', 'kinsman', 'heir', 'descendant', 'progeny', 'lineage', 'parentage']:
                keyword = 'family'

            window = 30
            person_before = None
            person_after = None
            for j in range(max(0, i - window), i):
                if tagged_tokens[j][1] == 'B-PER':
                    person_before = tagged_tokens[j][0]
                    break
            for j in range(i + 1, min(len(tagged_tokens), i + window + 1)):
                if tagged_tokens[j][1] == 'B-PER':
                    person_after = tagged_tokens[j][0]
                    break

            person = person_before or person_after
            if person:
                person = person.title()
                person_gender = gender_map.get(person, 'male')
                final_keyword = keyword
                valid_relation = False

                if keyword == 'son' and person_gender == 'male':
                    valid_relation = True
                elif keyword == 'daughter' and person_gender == 'female':
                    valid_relation = True
                elif keyword == 'father' and person_gender == 'male':
                    valid_relation = True
                elif keyword == 'mother' and person_gender == 'female':
                    valid_relation = True
                elif keyword == 'brother' and person_gender == 'male':
                    valid_relation = True
                elif keyword == 'sister' and person_gender == 'female':
                    valid_relation = True
                elif keyword in ['cousin', 'grandfather', 'grandsire', 'grandson', 'maternal uncle', 'brother-in-law', 'husband']:
                    valid_relation = True
                elif keyword == 'wife':
                    valid_relation = person_gender == 'female'
                elif keyword == 'family':
                    valid_relation = False

                if valid_relation and keyword in single_person_mappings and person in single_person_mappings[keyword]:
                    other_person, reverse_keyword = single_person_mappings[keyword][person]
                    other_gender = gender_map.get(other_person, 'male')
                    relation = (person, keyword, other_person)
                    if keyword == 'son' and other_gender not in ['female', 'male']:
                        valid_relation = False
                    elif keyword == 'daughter' and other_gender != 'male':
                        valid_relation = False
                    elif keyword == 'father' and other_gender not in ['male', 'female']:
                        valid_relation = False
                    elif keyword == 'mother' and other_gender not in ['male', 'female']:
                        valid_relation = False
                    elif keyword == 'brother' and other_gender != 'male':
                        valid_relation = False
                    elif keyword == 'sister' and other_gender != 'female':
                        valid_relation = False
                    elif keyword == 'wife' and other_gender != 'male':
                        valid_relation = False
                    if valid_relation and relation not in invalid_pairs:
                        potential_relations.append((relation, 'Valid (Single person)', person_gender, other_gender))
                        relations.append(relation)
                        reverse = (other_person, reverse_keyword, person)
                        if reverse not in invalid_pairs:
                            relations.append(reverse)
                    else:
                        potential_relations.append((relation, f'Invalid (Gender: {person_gender}, {other_gender})', person_gender, other_gender))
                else:
                    for pre_rel in predefined_relations:
                        if pre_rel[0] == person and pre_rel[1] == keyword and pre_rel[2] not in invalid_pairs:
                            relation = pre_rel
                            other_gender = gender_map.get(pre_rel[2], 'male')
                            potential_relations.append((relation, 'Valid (Predefined)', person_gender, other_gender))
                            relations.append(relation)
                            reverse = (pre_rel[2], single_person_mappings.get(keyword, {}).get(person, (None, None))[1] or keyword, person)
                            if reverse not in invalid_pairs:
                                relations.append(reverse)
                            break
                    else:
                        relation = (person, keyword, 'Unknown')
                        if relation not in invalid_pairs and keyword != 'family':
                            potential_relations.append((relation, 'Single person (Unmapped)', person_gender, 'N/A'))

            if person_before and person_after and person_before != person_after:
                person_before = person_before.title()
                person_after = person_after.title()
                before_gender = gender_map.get(person_before, 'male')
                after_gender = gender_map.get(person_after, 'male')
                valid_relation = False
                final_keyword = keyword
                relation = (person_before, final_keyword, person_after)

                if keyword == 'son':
                    if before_gender == 'male' and after_gender in ['female', 'male']:
                        valid_relation = True
                    elif after_gender == 'male' and before_gender == 'female':
                        valid_relation = True
                        person_before, person_after = person_after, person_before
                        final_keyword = 'mother'
                        relation = (person_before, final_keyword, person_after)
                elif keyword == 'daughter':
                    if before_gender == 'female' and after_gender == 'male':
                        valid_relation = True
                    elif after_gender == 'female' and before_gender == 'male':
                        valid_relation = True
                        person_before, person_after = person_after, person_before
                        final_keyword = 'father'
                        relation = (person_before, final_keyword, person_after)
                elif keyword == 'father':
                    if before_gender == 'male' and after_gender in ['male', 'female']:
                        valid_relation = True
                elif keyword == 'mother':
                    if before_gender == 'female' and after_gender in ['male', 'female']:
                        valid_relation = True
                elif keyword == 'brother':
                    if before_gender == 'male' and after_gender == 'male':
                        valid_relation = True
                elif keyword == 'sister':
                    if before_gender == 'female' and after_gender == 'female':
                        valid_relation = True
                elif keyword == 'cousin':
                    if before_gender == after_gender:
                        valid_relation = True
                elif keyword in ['grandfather', 'grandsire', 'grandson', 'maternal uncle', 'brother-in-law', 'husband']:
                    if before_gender == 'male':
                        valid_relation = True
                elif keyword == 'wife':
                    if before_gender == 'female' and after_gender == 'male':
                        valid_relation = True
                elif keyword == 'family':
                    valid_relation = (person_before, person_after) in predefined_family_pairs

                if valid_relation and relation not in invalid_pairs:
                    potential_relations.append((relation, 'Valid', before_gender, after_gender))
                    relations.append(relation)
                    if final_keyword == 'son' and before_gender == 'male' and after_gender == 'female':
                        reverse = (person_after, 'mother', person_before)
                        if reverse not in invalid_pairs:
                            relations.append(reverse)
                    elif final_keyword == 'daughter' and before_gender == 'female' and after_gender == 'male':
                        reverse = (person_after, 'father', person_before)
                        if reverse not in invalid_pairs:
                            relations.append(reverse)
                    elif final_keyword in ['brother', 'sister', 'cousin', 'brother-in-law', 'husband']:
                        reverse = (person_after, single_person_mappings.get(final_keyword, {}).get(person_before, (None, None))[1] or final_keyword, person_before)
                        if reverse not in invalid_pairs:
                            relations.append(reverse)
                    elif final_keyword == 'wife':
                        reverse = (person_after, 'husband', person_before)
                        if reverse not in invalid_pairs:
                            relations.append(reverse)
                    elif final_keyword == 'family':
                        reverse = (person_after, final_keyword, person_before)
                        if reverse not in invalid_pairs:
                            relations.append(reverse)
                else:
                    potential_relations.append((relation, f'Invalid (Gender: {before_gender}, {after_gender})', before_gender, after_gender))

            if not (person_before or person_after):
                start = max(0, i - 15)
                end = min(len(tagged_tokens), i + 16)
                context = ' '.join(t[0] for t in tagged_tokens[start:end])
                untagged_keywords.append((keyword, chapter, verse, context))

    for relation, status, gender1, gender2 in potential_relations:
        print(f"{status}: {relation[0]} -> {relation[1]} -> {relation[2]} (Gender: {gender1}, {gender2})")

    return relations, untagged_keywords

#Integrate all preprocessing and extraction steps for a single verse.
def process_verse(chapter, verse, text):
    text = clean_text(text)
    if not text:
        return [], [], []
    tokens = tokenize(text)
    tagged = tag_tokens(tokens)
    relations, untagged = extract_relations(tagged, chapter, verse)
    annotated = [(str(chapter), 'O'), ('.', 'O'), (str(verse), 'O')] + tagged
    # Orchestrates preprocessing and extraction, producing all necessary outputs for Cell 3.
    return annotated, relations, untagged

# Step 8: Load and process the CSV for Code 2
try:
    try:
        df = pd.read_csv("Bhagwad_Gita.csv", encoding='utf-8')
    except UnicodeDecodeError:
        df = pd.read_csv("Bhagwad_Gita.csv", encoding='latin1')
    gita_data = df[['Chapter', 'Verse', 'EngMeaning']].dropna()
    gita_data = gita_data.rename(columns={'EngMeaning': 'Text'})
except FileNotFoundError:
    print("Error: 'Bhagwad_Gita.csv' not found.")
    exit(1)
except Exception as e:
    print(f"Error loading CSV: {e}")
    exit(1)

# Log CSV details
print(f"Loaded CSV: {len(gita_data)} rows")
print(f"Columns found: {list(gita_data.columns)}")
print(f"First 5 rows:\n{gita_data.head(5).to_string()}")
print(f"First row: {gita_data.iloc[0][['Chapter', 'Verse', 'Text']].to_dict()}")

print("Setup complete. Run the next cell to process the data and extract relations.")
```

```python
# Cell 2: Process all verses, extract relations, and prepare train/dev/test splits

import random
from itertools import combinations

# Process all verses and collect relations
all_tagged = []
all_relations = []
all_untagged_keywords = []
verse_relations = []
verse_tagged = []

for idx, row in gita_data.iterrows():
    chapter, verse, text = row['Chapter'], row['Verse'], row['Text']
    annotated, relations, untagged = process_verse(chapter, verse, text)
    verse_tagged.append((idx, annotated))
    verse_relations.append((idx, relations))
    all_untagged_keywords.extend(untagged)

# Normalize and deduplicate relations
all_relations_set = set()
for idx, relations in verse_relations:
    for rel in relations:
        e1 = name_to_canonical.get(rel[0].lower(), rel[0].lower())
        e2 = name_to_canonical.get(rel[2].lower(), rel[2].lower())
        rel_norm = (e1, rel[1], e2)
        if rel_norm not in invalid_pairs:
            all_relations_set.add(rel_norm)

# Add all predefined relations
for rel in predefined_relations:
    e1 = name_to_canonical.get(rel[0].lower(), rel[0].lower())
    e2 = name_to_canonical.get(rel[2].lower(), rel[2].lower())
    rel_norm = (e1, rel[1], e2)
    if rel_norm not in invalid_pairs:
        all_relations_set.add(rel_norm)

all_relations_list = list(all_relations_set)

# Generate negative samples (non-existent relations)
negative_relations = []
possible_pairs = list(combinations(person_names, 2))
random.shuffle(possible_pairs)
neg_count = min(len(all_relations_list), len(possible_pairs))  # Equal number of positive and negative
for p1, p2 in possible_pairs[:neg_count]:
    for rel in ['son', 'father', 'mother', 'brother', 'sister', 'cousin', 'husband', 'wife']:
        rel_tuple = (p1.lower(), rel, p2.lower())
        if (rel_tuple not in all_relations_set and
            rel_tuple not in invalid_pairs and
            gender_map.get(p1, 'male') != 'unknown' and
            gender_map.get(p2, 'male') != 'unknown'):
            negative_relations.append(rel_tuple)
            break  # Add one negative relation per pair

all_relations_list.extend(negative_relations)

# Split relations into train/dev/test (70:15:15)
train_rels, temp_rels = train_test_split(all_relations_list, test_size=0.30, random_state=42)
dev_rels, test_rels = train_test_split(temp_rels, test_size=0.5, random_state=42)

train_verses = []
dev_verses = []
test_verses = []
train_relations = []
dev_relations = []
test_relations = []

# Assign verses to splits based on relations
for idx, relations in verse_relations:
    verse_has_train = False
    verse_has_dev = False
    verse_has_test = False
    for rel in relations:
        e1 = name_to_canonical.get(rel[0].lower(), rel[0].lower())
        e2 = name_to_canonical.get(rel[2].lower(), rel[2].lower())
        rel_norm = (e1, rel[1], e2)
        if rel_norm in train_rels and not verse_has_train:
            train_verses.append((idx, verse_tagged[idx][1], [rel_norm]))
            train_relations.append(rel_norm)
            verse_has_train = True
        elif rel_norm in dev_rels and not verse_has_dev:
            dev_verses.append((idx, verse_tagged[idx][1], [rel_norm]))
            dev_relations.append(rel_norm)
            verse_has_dev = True
        elif rel_norm in test_rels and not verse_has_test:
            test_verses.append((idx, verse_tagged[idx][1], [rel_norm]))
            test_relations.append(rel_norm)
            verse_has_test = True

# Ensure predefined relations are included
for rel in predefined_relations:
    e1 = name_to_canonical.get(rel[0].lower(), rel[0].lower())
    e2 = name_to_canonical.get(rel[2].lower(), rel[2].lower())
    rel_norm = (e1, rel[1], e2)
    if rel_norm in train_rels and rel_norm not in train_relations:
        train_relations.append(rel_norm)
    elif rel_norm in dev_rels and rel_norm not in dev_relations:
        dev_relations.append(rel_norm)
    elif rel_norm in test_rels and rel_norm not in test_relations:
        test_relations.append(rel_norm)

# Add negative relations to splits
for rel in negative_relations:
    if rel in train_rels and rel not in train_relations:
        train_relations.append(rel)
    elif rel in dev_rels and rel not in dev_relations:
        dev_relations.append(rel)
    elif rel in test_rels and rel not in test_relations:
        test_relations.append(rel)

# Handle remaining verses
remaining_verses = [(idx, tagged) for idx, tagged in verse_tagged if idx not in [v[0] for v in train_verses + dev_verses + test_verses]]
train_rem, temp_rem = train_test_split(remaining_verses, test_size=0.30, random_state=42)
dev_rem, test_rem = train_test_split(temp_rem, test_size=0.5, random_state=42)

train_verses.extend([(idx, tagged, []) for idx, tagged in train_rem])
dev_verses.extend([(idx, tagged, []) for idx, tagged in dev_rem])
test_verses.extend([(idx, tagged, []) for idx, tagged in test_rem])

train_verses = sorted(train_verses, key=lambda x: x[0])
dev_verses = sorted(dev_verses, key=lambda x: x[0])
test_verses = sorted(test_verses, key=lambda x: x[0])

# Write output files
def write_tagged_file(filename, verses):
    with open(filename, 'w', encoding='utf-8') as f:
        for _, tagged, _ in verses:
            for word, tag in tagged:
                if tag not in ['O', 'B-PER', 'I-PER', 'B-LOC', 'I-LOC', 'B-REL', 'I-REL']:
                    print(f"Warning: Invalid tag '{tag}' for word '{word}' in {filename}")
                    tag = 'O'
                f.write(f"{word} {tag}\n")
            f.write("\n")

def write_relations_file(filename, relations):
    seen = set()
    unique_relations = []
    for rel in relations:
        rel_tuple = (rel[0], rel[1], rel[2])
        if rel_tuple not in seen and rel_tuple not in invalid_pairs:
            seen.add(rel_tuple)
            unique_relations.append(rel)
    with open(filename, 'w', encoding='utf-8') as f:
        for subj, rel, obj in unique_relations:
            f.write(f"{subj} -> {rel} -> {obj}\n")

# Verify Google Drive mount and create directory
save_dir = "/content/drive/My Drive/Gita_Pipeline_Output"
if not os.path.exists('/content/drive'):
    raise RuntimeError("Google Drive is not mounted. Please mount Google Drive and try again.")
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    print(f"Created directory: {save_dir}")

# Save files to Google Drive
write_tagged_file(os.path.join(save_dir, "train.txt"), train_verses)
write_tagged_file(os.path.join(save_dir, "dev.txt"), dev_verses)
write_tagged_file(os.path.join(save_dir, "test.txt"), test_verses)
write_relations_file(os.path.join(save_dir, "train_relations.txt"), train_relations)
write_relations_file(os.path.join(save_dir, "dev_relations.txt"), dev_relations)
write_relations_file(os.path.join(save_dir, "test_relations.txt"), test_relations)

# Enhanced logging
print(f"Total unique relations: {len(all_relations_list)} (Positive: {len(all_relations_set)}, Negative: {len(negative_relations)})")
print(f"Train relations: {len(set(train_relations))} (sample: {train_relations[:5]})")
print(f"Dev relations: {len(set(dev_relations))} (sample: {dev_relations[:5]})")
print(f"Test relations: {len(set(test_relations))} (sample: {test_relations[:5]})")

print("\nRelation type distribution (Train):")
relation_counts = defaultdict(int)
for _, rel, _ in train_relations:
    relation_counts[rel] += 1
print(dict(relation_counts))

print("\nSample Tagged Verses (Train, first 2):")
for _, tagged, _ in train_verses[:2]:
    print("\n".join(f"{word} {tag}" for word, tag in tagged))
    print()

print("\nSample Extracted Relations (Train, first 5 non-predefined):")
extracted = [r for r in train_relations if r not in predefined_relations][:5]
for rel in extracted:
    print(f"{rel[0]} -> {rel[1]} -> {rel[2]}")
print("Files saved to Google Drive. Run the next cell to build and visualize the graph.")
```

```python
# Cell 3: Build and Visualize the Graph

# Step 9: Load the dataset
df = pd.read_csv("Bhagwad_Gita.csv")
eng_meanings = df["EngMeaning"]

# Step 10-12: Tokenization, NER, and Relation Extraction (unchanged)
def tokenize_and_remove_stopwords(text):
    tokens = []
    text = clean_text(text)
    doc = nlp(text)
    i = 0
    while i < len(doc):
        matched = False
        for j in range(3, 0, -1):
            if i + j <= len(doc):
                candidate = ' '.join(t.text for t in doc[i:i+j])
                candidate_lower = candidate.lower()
                if (candidate_lower in person_names_set or
                    candidate_lower in locations_set or
                    candidate_lower in relation_keywords_set):
                    tokens.append(candidate)
                    i += j
                    matched = True
                    break
        if not matched:
            token = doc[i]
            token_lower = token.text.lower()
            if (not token.is_stop or
                token_lower in person_names_set or
                token_lower in locations_set or
                token_lower in relation_keywords_set):
                tokens.append(token.text)
            i += 1
    return [t for t in tokens if t]

def tag_tokens_gnn(tokens):
    tagged = []
    i = 0
    while i < len(tokens):
        token = tokens[i]
        token_lower = token.lower()
        if token_lower in person_names_set:
            tagged.append((token, 'B-PER'))
        elif token_lower in locations_set:
            tagged.append((token, 'B-LOC'))
        elif token_lower in relation_keywords_set:
            tagged.append((token, 'B-REL'))
            j = i + 1
            while j < len(tokens):
                combined = ' '.join(tokens[i:j+1]).lower()
                if combined in relation_keywords_set:
                    tagged.append((tokens[j], 'I-REL'))
                    j += 1
                else:
                    break
            i = j - 1
        else:
            tagged.append((token, 'O'))
        i += 1
    return tagged

def extract_relations_gnn(tagged_tokens):
    relations = []
    for i, (token, tag) in enumerate(tagged_tokens):
        if tag == 'B-REL':
            keyword = token.lower()
            if keyword in ['sons', 'child', 'children', 'offspring', 'son of', 'children of', 'child of', 'born', 'scion', 'issue', 'begotten']:
                keyword = 'son'
            elif keyword in ['daughters', 'daughter of']:
                keyword = 'daughter'
            elif keyword in ['fathers', 'parents', 'ancestors', 'forefather', 'progenitor', 'sire', 'ancestry', 'heritage', 'clan', 'tribe', 'household']:
                keyword = 'father'
            elif keyword in ['brothers', 'siblings', 'brethren']:
                keyword = 'brother'
            elif keyword in ['sisters']:
                keyword = 'sister'
            elif keyword in ['cousins']:
                keyword = 'cousin'
            elif keyword in ['grandfathers', 'grandsons', 'maternal uncles', 'descendants']:
                keyword = keyword.rstrip('s')
            elif keyword in ['kin', 'kinsmen', 'relatives', 'born to', 'born of', 'relative', 'kinsman', 'heir', 'descendant', 'progeny', 'lineage', 'parentage']:
                keyword = 'family'

            window = 30
            person_before = None
            person_after = None
            for j in range(max(0, i - window), i):
                if tagged_tokens[j][1] == 'B-PER':
                    person_before = tagged_tokens[j][0]
                    break
            for j in range(i + 1, min(len(tagged_tokens), i + window + 1)):
                if tagged_tokens[j][1] == 'B-PER':
                    person_after = tagged_tokens[j][0]
                    break

            if person_before and person_after and person_before != person_after:
                person_before = person_before.lower()
                person_after = person_after.lower()
                person_before = name_to_canonical.get(person_before, person_before)
                person_after = name_to_canonical.get(person_after, person_after)
                before_gender = gender_map.get(person_before, 'unknown')
                after_gender = gender_map.get(person_after, 'unknown')
                valid_relation = False

                if keyword == 'son' and before_gender == 'male' and after_gender in ['male', 'female']:
                    valid_relation = True
                elif keyword == 'daughter' and before_gender == 'female' and after_gender == 'male':
                    valid_relation = True
                elif keyword == 'father' and before_gender == 'male' and after_gender in ['male', 'female']:
                    valid_relation = True
                elif keyword == 'mother' and before_gender == 'female' and after_gender in ['male', 'female']:
                    valid_relation = True
                elif keyword == 'brother' and before_gender == 'male' and after_gender == 'male':
                    valid_relation = True
                elif keyword == 'sister' and before_gender == 'female' and after_gender == 'female':
                    valid_relation = True
                elif keyword == 'cousin':
                    valid_relation = True
                elif keyword in ['grandfather', 'grandson', 'maternal uncle', 'brother-in-law', 'husband'] and before_gender == 'male':
                    valid_relation = True
                elif keyword == 'wife' and before_gender == 'female' and after_gender == 'male':
                    valid_relation = True
                elif keyword == 'family':
                    valid_relation = True

                if valid_relation and (person_before, keyword, person_after) not in invalid_pairs:
                    relations.append((person_before, keyword, person_after))

    return relations

# Step 13: Load relations
def load_relations_from_file(filename):
    relations = []
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if line:
                parts = line.split(" -> ")
                if len(parts) == 3:
                    person1, rel, person2 = parts
                    person1 = person1.lower()
                    person2 = person2.lower()
                    person1 = name_to_canonical.get(person1, person1)
                    person2 = name_to_canonical.get(person2, person2)
                    relations.append((person1, rel, person2))
    return relations

save_dir = "/content/drive/My Drive/Gita_Pipeline_Output"
train_relations = load_relations_from_file(os.path.join(save_dir, "train_relations_cleaned.txt")) #train_relations_cleaned.txt
dev_relations = load_relations_from_file(os.path.join(save_dir, "dev_relations_cleaned.txt")) #dev_relations_cleaned.txt
test_relations = load_relations_from_file(os.path.join(save_dir, "test_relations_cleaned.txt")) #test_relations_cleaned.txt
all_relations = train_relations + dev_relations + test_relations

# Step 14: Graph Construction with Verification
G = nx.DiGraph()
# Add nodes with gender
for char in person_names:
    if char in person_names_set:  # Ensure only valid persons are nodes
        G.add_node(char, gender=gender_map.get(char, "unknown"))

# Add manual relations
manual_relations = []
for person1, rel, person2 in manual_relations:
    if person1 in person_names and person2 in person_names:
        G.add_edge(person1, person2, relation=rel)
# Add extracted relations with strict validation
invalid_entities = {
    'mrityu', 'death', 'god', 'lord', 'divinity', 'heaven', 'soul', 'spirit', 'war', 'battle', 'chariot', 'divine',
    'fate', 'destiny', 'earth', 'sky', 'fire', 'wind', 'water', 'time', 'veda', 'scripture', 'knowledge', 'unknown'
}
for person1, rel, person2 in all_relations:
    if (person1 in person_names and person2 in person_names and
        person1 not in invalid_entities and person2 not in invalid_entities and
        (person1, rel, person2) not in invalid_pairs):
        p1_gender = gender_map.get(person1, 'unknown')
        p2_gender = gender_map.get(person2, 'unknown')
        valid = True
        if rel == 'son' and (p1_gender != 'male' or p2_gender == 'female'):
            valid = False
        elif rel == 'daughter' and (p1_gender != 'female' or p2_gender != 'male'):
            valid = False
        elif rel == 'father' and p1_gender != 'male':
            valid = False
        elif rel == 'mother' and p1_gender != 'female':
            valid = False
        elif rel == 'brother' and (p1_gender != 'male' or p2_gender != 'male'):
            valid = False
        elif rel == 'sister' and (p1_gender != 'female' or p2_gender != 'female'):
            valid = False
        elif rel == 'husband' and (p1_gender != 'male' or p2_gender != 'female'):
            valid = False
        elif rel == 'wife' and (p1_gender != 'female' or p2_gender != 'male'):
            valid = False
        elif rel == 'friend' and p1_gender == 'unknown':
            valid = False
        if valid and not G.has_edge(person1, person2):
            G.add_edge(person1, person2, relation=rel)

# Extract relations from text and add to graph
for meaning in eng_meanings:
    tokens = tokenize_and_remove_stopwords(meaning)
    tagged = tag_tokens_gnn(tokens)
    relations = extract_relations_gnn(tagged)
    for person1, rel, person2 in relations:
        if (person1 in person_names and person2 in person_names and
            person1 not in invalid_entities and person2 not in invalid_entities and
            (person1, rel, person2) not in invalid_pairs):
            p1_gender = gender_map.get(person1, 'unknown')
            p2_gender = gender_map.get(person2, 'unknown')
            valid = True
            if rel == 'son' and (p1_gender != 'male' or p2_gender == 'female'):
                valid = False
            elif rel == 'daughter' and (p1_gender != 'female' or p2_gender != 'male'):
                valid = False
            elif rel == 'father' and p1_gender != 'male':
                valid = False
        elif rel == 'mother' and p1_gender != 'female':
            valid = False
        elif rel == 'brother' and (p1_gender != 'male' or p2_gender != 'male'):
            valid = False
        elif rel == 'sister' and (p1_gender != 'female' or p2_gender != 'female'):
            valid = False
        elif rel == 'husband' and (p1_gender != 'male' or p2_gender != 'female'):
            valid = False
        elif rel == 'wife' and (p1_gender != 'female' or p2_gender != 'male'):
            valid = False
        elif rel == 'friend' and p1_gender == 'unknown':
            valid = False
        if valid and not G.has_edge(person1, person2):
            G.add_edge(person1, person2, relation=rel)
            if rel in ["son", "daughter"]:
                reverse_rel = "father" if rel == "son" else "mother"
                if not G.has_edge(person2, person1) and (person2, reverse_rel, person1) not in invalid_pairs:
                    G.add_edge(person2, person1, relation=reverse_rel)
            if rel in ["brother", "sister", "cousin", "brother-in-law"]:
                if not G.has_edge(person2, person1) and (person2, rel, person1) not in invalid_pairs:
                    G.add_edge(person2, person1, relation=rel)
            if rel == "husband":
                if not G.has_edge(person2, person1) and (person2, "wife", person1) not in invalid_pairs:
                    G.add_edge(person2, person1, relation="wife")
            if rel == "wife":
                if not G.has_edge(person2, person1) and (person2, "husband", person1) not in invalid_pairs:
                    G.add_edge(person2, person1, relation="husband")

# Enhanced edge validation
invalid_edges = []
for u, v, data in list(G.edges(data=True)):
    rel = data['relation']
    p1_gender = gender_map.get(u, 'unknown')
    p2_gender = gender_map.get(v, 'unknown')

    # Check for invalid entities
    if u in invalid_entities or v in invalid_entities:
        invalid_edges.append((u, v, rel))
        continue

    # Check invalid pairs
    if (u, rel, v) in invalid_pairs:
        invalid_edges.append((u, v, rel))
        continue

    # Minimal specific checks for edge cases not in invalid_pairs
    if rel in ['brother', 'sister', 'cousin', 'brother-in-law'] and not G.has_edge(v, u):
        invalid_edges.append((u, v, rel))
        continue
    if rel == 'husband' and not any(d.get('relation') == 'wife' for _, _, d in G.edges(v, data=True)):
        invalid_edges.append((u, v, rel))
        continue
    if rel == 'wife' and not any(d.get('relation') == 'husband' for _, _, d in G.edges(v, data=True)):
        invalid_edges.append((u, v, rel))
        continue
    if rel in ['son', 'daughter']:
        reverse_rel = 'father' if rel == 'son' else 'mother'
        if not any(d.get('relation') == reverse_rel for _, _, d in G.edges(v, data=True)):
            invalid_edges.append((u, v, rel))
            continue

    # Degree-based filtering
    if G.degree(u) > 20 or G.degree(v) > 20:  # Adjust threshold based on data
        invalid_edges.append((u, v, rel))
        continue

# Remove invalid edges
for u, v, rel in invalid_edges:
    if G.has_edge(u, v):
        G.remove_edge(u, v)

# Verify graph edges
key_characters = ['krishna', 'arjuna', 'bhima', 'draupadi', 'kunti', 'pandu', 'dhritarashtra', 'subhadra', 'vasudeva', 'karna', 'duryodhana', 'balarama']
print("Graph edges for key characters:")
for char in key_characters:
    print(f"\nEdges involving {char}:")
    edges_found = False
    for u, v, data in G.edges(data=True):
        if u == char or v == char:
            edge_str = f"{u} -> {data['relation']} -> {v}"
            print(edge_str)
            edges_found = True
    if not edges_found:
        print("No edges found.")

# Log invalid edges if any
if invalid_edges:
    print("\nInvalid edges removed:")
    for u, v, rel in invalid_edges:
        print(f"{u} -> {rel} -> {v}")
else:
    print("\nNo invalid edges found.")

# Log graph stats
print("\nGraph stats:")
print(f"Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")

# Save the graph
with open(os.path.join(save_dir, "gita_graph.pkl"), "wb") as f:
    pickle.dump(G, f)
nx.write_gml(G, os.path.join(save_dir, "family_tree.gml"))

# Step 16: Visualize the Graph
def visualize_graph(G, focus_nodes=None):
    net = Network(height="750px", width="100%", notebook=True, directed=True, cdn_resources='remote')
    if focus_nodes:
        subgraph = G.subgraph(focus_nodes)
    else:
        subgraph = G

    for node in subgraph.nodes:
        net.add_node(node, label=node, color="lightblue" if gender_map.get(node, "unknown") == "male" else "pink")

    for u, v, data in subgraph.edges(data=True):
        net.add_edge(u, v, label=data["relation"], color="gray")

    visualization_file = os.path.join(save_dir, "family_tree.html")
    net.show(visualization_file)
    return visualization_file

# Visualize both
focus_nodes = ["arjuna", "krishna", "bhima", "kunti", "pandu", "dhritarashtra", "vasudeva", "karna", "duryodhana"]
visualization_file = visualize_graph(G, focus_nodes)
print(f"Graph visualization saved to {visualization_file}")
print(f"Graph saved as {os.path.join(save_dir, 'gita_graph.pkl')} and {os.path.join(save_dir, 'family_tree.gml')}")

print("Graph built and visualized. Run the next cell to train the GNN model.")
```

### Train the GNN model

```python
# Cell 4: Train the GNN Model
# Step 16: Precompute Mention Counts
mention_counts = defaultdict(int)
for meaning in eng_meanings:
    tokens = tokenize_and_remove_stopwords(meaning)
    for token in tokens:
        token_lower = token.lower()
        canonical_name = name_to_canonical.get(token_lower, token_lower)
        if canonical_name in person_names_set:
            mention_counts[canonical_name] += 1

# Step 17: Prepare Data for GNN
node_list = list(G.nodes())
node2idx = {node: idx for idx, node in enumerate(node_list)}
num_nodes = len(node_list)

# Node features with learned embeddings
node_features = []
for node in node_list:
    degree = G.degree(node)
    mention_count = mention_counts[node]
    in_degree = G.in_degree(node)
    out_degree = G.out_degree(node)
    clustering = nx.clustering(G.to_undirected(), node)
    gender = 1.0 if gender_map.get(node, 'unknown') == 'male' else 0.0 if gender_map.get(node, 'unknown') == 'female' else 0.5
    node_features.append([degree, mention_count, in_degree, out_degree, clustering, gender])

# Normalize features
node_features = np.array(node_features)
node_features[:, :-1] = (node_features[:, :-1] - node_features[:, :-1].mean(axis=0)) / (node_features[:, :-1].std(axis=0) + 1e-8)
x = torch.tensor(node_features, dtype=torch.float).to(device)

# Learned node embeddings
embed_dim = 32
char_embedding = torch.nn.Embedding(num_nodes, embed_dim).to(device)
char_ids = torch.arange(num_nodes).to(device)
char_embed = char_embedding(char_ids)
x = torch.cat([x, char_embed], dim=1)  # Append 32-dim embeddings

edges = []
edge_labels = []
relation_types = list(set([data["relation"] for _, _, data in G.edges(data=True)]))
relation_encoder = LabelEncoder()
relation_encoder.fit(relation_types)

for u, v, data in G.edges(data=True):
    edges.append([node2idx[u], node2idx[v]])
    edge_labels.append(relation_encoder.transform([data["relation"]])[0])

edges = torch.tensor(edges, dtype=torch.long).t().contiguous().to(device)
edge_labels = torch.tensor(edge_labels, dtype=torch.long).to(device)

# Class weights
unique_labels, label_counts = np.unique(edge_labels.cpu().numpy(), return_counts=True)
class_weights = 1.0 / (label_counts + 1e-8) ** 0.5
class_weights = class_weights / class_weights.sum() * len(unique_labels)
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

data = Data(x=x, edge_index=edges, edge_attr=edge_labels, num_nodes=num_nodes).to(device)

# Train/dev/test masks
edge_indices = torch.arange(edges.shape[1])
train_mask = torch.zeros(edges.shape[1], dtype=torch.bool).to(device)
dev_mask = torch.zeros(edges.shape[1], dtype=torch.bool).to(device)
test_mask = torch.zeros(edges.shape[1], dtype=torch.bool).to(device)

for i, (u, v) in enumerate(edges.t()):
    u_node, v_node = node_list[u], node_list[v]
    relation = G.get_edge_data(u_node, v_node)["relation"]
    rel_tuple = (u_node, relation, v_node)
    if rel_tuple in train_relations:
        train_mask[i] = 1
    elif rel_tuple in dev_relations:
        dev_mask[i] = 1
    elif rel_tuple in test_relations:
        test_mask[i] = 1

data.train_mask = train_mask
data.dev_mask = dev_mask
data.test_mask = test_mask

# Step 18: Define GCN Model
class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.bn1 = torch.nn.BatchNorm1d(hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.bn2 = torch.nn.BatchNorm1d(hidden_channels)
        self.conv3 = GCNConv(hidden_channels, out_channels)
        self.bn3 = torch.nn.BatchNorm1d(out_channels)
        self.dropout = torch.nn.Dropout(0.5)
        self.fc = torch.nn.Linear(out_channels * 2, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.conv2(x, edge_index)
        x = self.bn2(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.conv3(x, edge_index)
        x = self.bn3(x)
        x = F.relu(x)
        x = self.dropout(x)
        edge_embeddings = torch.cat([x[edge_index[0]], x[edge_index[1]]], dim=1)
        out = self.fc(edge_embeddings)
        return out

# Initialize model
num_classes = len(relation_types)
in_channels = x.shape[1]  # 6 features + 32-dim embedding
model = GCN(in_channels=in_channels, hidden_channels=256, out_channels=128, num_classes=num_classes).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=5e-4)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=300, eta_min=1e-6)

# Label smoothing loss
class LabelSmoothingLoss(torch.nn.Module):
    def __init__(self, smoothing=0.1):
        super(LabelSmoothingLoss, self).__init__()
        self.smoothing = smoothing
        self.cls = num_classes

    def forward(self, pred, target):
        pred = pred.log_softmax(dim=-1)
        with torch.no_grad():
            true_dist = torch.zeros_like(pred)
            true_dist.fill_(self.smoothing / (self.cls - 1))
            true_dist.scatter_(1, target.unsqueeze(1), 1.0 - self.smoothing)
        return torch.mean(torch.sum(-true_dist * pred, dim=-1))

criterion = LabelSmoothingLoss(smoothing=0.1)

# Training and evaluation
def train():
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = criterion(out[data.train_mask], data.edge_attr[data.train_mask])
    loss.backward(retain_graph=True)
    optimizer.step()
    scheduler.step()
    return loss.item()

def evaluate(mask, log_misclassifications=False, epoch=None):
    model.eval()
    misclassifications = []
    with torch.no_grad():
        out = model(data)
        pred = out.argmax(dim=1)
        correct = (pred[mask] == data.edge_attr[mask]).sum()
        acc = int(correct) / int(mask.sum()) if mask.sum() > 0 else 0
        if log_misclassifications:
            incorrect = (pred[mask] != data.edge_attr[mask]).nonzero(as_tuple=True)[0]
            for idx in incorrect:
                edge_idx = torch.where(mask)[0][idx]
                u, v = edges[:, edge_idx]
                true_rel = relation_encoder.inverse_transform([data.edge_attr[edge_idx].item()])[0]
                pred_rel = relation_encoder.inverse_transform([pred[edge_idx].item()])[0]
                misclass = f"{node_list[u]} -> {true_rel} -> {node_list[v]} (Predicted: {pred_rel})"
                misclassifications.append(misclass)
    return acc, misclassifications

# Training loop
best_dev_acc = 0
best_epoch = 0
patience = 100
wait = 0
misclassification_log = []

# Log dataset sizes
print(f"Train edges: {data.train_mask.sum().item()}, Dev edges: {data.dev_mask.sum().item()}, Test edges: {data.test_mask.sum().item()}")

for epoch in range(500):
    loss = train()
    train_acc, _ = evaluate(data.train_mask)
    dev_acc, dev_misclass = evaluate(data.dev_mask, log_misclassifications=True, epoch=epoch)
    test_acc, _ = evaluate(data.test_mask)
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}, Train Acc: {train_acc:.4f}, Dev Acc: {dev_acc:.4f}, Test Acc: {test_acc:.4f}")
        if dev_misclass:
            print("Dev Misclassifications:")
            for misclass in dev_misclass[:5]:
                print(misclass)
        misclassification_log.extend([(epoch, misclass) for misclass in dev_misclass])

    if dev_acc > best_dev_acc:
        best_dev_acc = dev_acc
        best_epoch = epoch
        wait = 0
        torch.save({
            'model_state_dict': model.state_dict(),
            'node2idx': node2idx,
            'node_list': node_list,
            'num_nodes': num_nodes,
            'num_classes': num_classes,
            'node_features': x
        }, os.path.join(save_dir, 'gita_gnn_model.pt'))
    else:
        wait += 1
        if wait >= patience:
            print(f"Early stopping at epoch {epoch}. Best dev accuracy: {best_dev_acc:.4f} at epoch {best_epoch}")
            break

# Save misclassifications
with open(os.path.join(save_dir, 'misclassifications.txt'), 'w') as f:
    for epoch, misclass in misclassification_log:
        f.write(f"Epoch {epoch}: {misclass}\n")

# Save model and encoder
with open(os.path.join(save_dir, 'relation_encoder_classes.pkl'), 'wb') as f:
    pickle.dump(relation_encoder.classes_, f)

print(f"Model saved as {os.path.join(save_dir, 'gita_gnn_model.pt')}")
print(f"Relation encoder saved as {os.path.join(save_dir, 'relation_encoder_classes.pkl')}")
print(f"Misclassifications saved as {os.path.join(save_dir, 'misclassifications.txt')}")

print("GNN model trained. Run the next cell to predict relationships.")
```

### Cell 5: Predict Relationships

```python
# Cell 5: Predict Relationships

# Step 19: Test the Model with Relationship Inference
def predict_relationship(model, person1, person2, node2idx, relation_encoder, G, reverse=False):
    person1 = name_to_canonical.get(person1.lower(), person1.lower())
    person2 = name_to_canonical.get(person2.lower(), person2.lower())

    if person1 not in node2idx or person2 not in node2idx:
        return f"No relationship found between {person1} and {person2}", 0.0

    reverse_relations = {
        "father": "son", "mother": "daughter", "son": "father", "daughter": "mother",
        "grandfather": "grandson", "grandson": "grandfather", "brother": "brother",
        "sister": "sister", "cousin": "cousin", "brother-in-law": "brother-in-law",
        "sister-in-law": "sister-in-law", "husband": "wife", "wife": "husband",
        "maternal uncle": "nephew/niece", "father-in-law": "son-in-law", "family": "family"
    }

    if not reverse and G.has_edge(person1, person2):
        relation = G.get_edge_data(person1, person2)["relation"]
        return f"{person1} is the {relation} of {person2}", 1.0
    elif reverse and G.has_edge(person2, person1):
        relation = G.get_edge_data(person2, person1)["relation"]
        reverse_relation = reverse_relations.get(relation, "unknown relation")
        return f"{person1} is the {reverse_relation} of {person2}", 1.0

    idx1, idx2 = node2idx[person1], node2idx[person2]
    inference_edge = torch.tensor([[idx1], [idx2]], dtype=torch.long).to(device)
    inference_data = Data(x=data.x, edge_index=inference_edge).to(device)

    model.eval()
    with torch.no_grad():
        out = model(inference_data)
        probs = F.softmax(out, dim=1)
        pred = out.argmax(dim=1)
        confidence = probs[0, pred[0]].item()
        relation = relation_encoder.inverse_transform(pred.cpu().numpy())[0]
    if reverse:
        relation = reverse_relations.get(relation, "unknown relation")
    return f"Predicted: {person1} is the {relation} of {person2}", confidence

def determine_relationships(user_input, model, node2idx, relation_encoder, G):
    names = [name.strip().lower() for name in user_input.replace(" ", ",").split(",") if name.strip()]
    if len(names) < 2:
        return [("Please provide at least two names (e.g., arjuna, krishna).", 0.0)]

    first_person = names[0]
    other_persons = names[1:]
    results = []
    for person in other_persons:
        relationship, confidence = predict_relationship(model, first_person, person, node2idx, relation_encoder, G, reverse=False)
        if "No relationship found" in relationship or "Predicted" in relationship:
            relationship, confidence = predict_relationship(model, first_person, person, node2idx, relation_encoder, G, reverse=True)
        results.append((relationship, confidence))
    return results

# Step 20: Test Specific Relationships
test_pairs = [
    ("krishna", "bhima"),  # Should be cousin
    ("arjuna", "draupadi"),  # Should be husband
    ("kunti", "arjuna"),  # Should be mother
    ("dhritarashtra", "pandu"),  # Should be brother
    ("subhadra", "krishna")  # Should be sister
]

print("Testing key relationships:")
for person1, person2 in test_pairs:
    relationship, confidence = predict_relationship(model, person1, person2, node2idx, relation_encoder, G, reverse=False)
    print(f"{relationship} (Confidence: {confidence:.2f})")
    if "Predicted" in relationship:
        relationship, confidence = predict_relationship(model, person1, person2, node2idx, relation_encoder, G, reverse=True)
        print(f"Reverse: {relationship} (Confidence: {confidence:.2f})")

# Interactive testing
print("\nEnter names separated by commas (e.g., arjuna, krishna, draupadi):")
user_input = input()
relationships = determine_relationships(user_input, model, node2idx, relation_encoder, G)
for rel, conf in relationships:
    print(f"{rel} (Confidence: {conf:.2f})")

print(f"All output files are saved in {save_dir}. You can access them from your Google Drive.")
```
